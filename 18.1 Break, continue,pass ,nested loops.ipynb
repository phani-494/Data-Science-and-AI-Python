# break

## Purpose of Transfer Control Statements in Python 🔄

### 1️⃣ Concept

**Transfer control statements** are used to **change the normal flow of execution inside loops or blocks of code**.
To transfer the control of PVM from one part of the program to another part of the program.

Normally, a loop executes **sequentially** from top to bottom.
But sometimes we need to:

* stop the loop
* skip an iteration
* do nothing temporarily

For this purpose we use **transfer control statements**.

---

# 2️⃣ Types of Transfer Control Statements

In Python there are **three main transfer control statements**:

| Statement  | Purpose                        |
| ---------- | ------------------------------ |
| `break`    | Terminates the loop completely |
| `continue` | Skips current iteration        |
| `pass`     | Does nothing (placeholder)     |
| `return`   | terminates from function      |

---

# 3️⃣ `break` Statement

### Purpose

`break` **stops the loop immediately** and transfers control outside the loop.

### Syntax

```python
for/while condition:
    if condition:
        break
```

### Example

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

Output

```
0
1
2
3
4
```

Explanation
When `i = 5`, the loop **stops completely**.

---

# 4️⃣ `continue` Statement

### Purpose

`continue` **skips the current iteration** and moves to the next iteration.

### Syntax

```python
for/while condition:
    if condition:
        continue
```

### Example

```python
for i in range(5):
    if i == 2:
        continue
    print(i)
```

Output

```
0
1
3
4
```

Explanation
When `i = 2`, Python **skips that iteration**.

---

# 5️⃣ `pass` Statement

### Purpose

`pass` **does nothing**. It is used as a **placeholder** when a statement is required syntactically but no action is needed.

### Example

```python
for i in range(5):
    pass
```

Program runs but **nothing happens**.
