## Data Types 

# 🔹 What is the Purpose of Data Types in Python?

Data types tell the computer:

> **What kind of data is being stored and what operations are allowed on it.**



## 🎯 Main Purposes of Data Types

## 1️⃣ Identify the Type of Data

The computer must know whether the value is:

* A number → `int`, `float`
* Text → `str`
* Collection → `list`, `tuple`, `dict`
* Logical value → `bool`

Example:

```python
x = 10        # int
name = "AI"   # string
```

If Python doesn't know the type, it cannot process the value correctly.

---

## 2️⃣ Decide What Operations Are Allowed

Each data type supports different operations.

Example:


10 + 5        # ✔ Allowed (numbers)
"AI" + "ML"   # ✔ Allowed (string concatenation)
10 + "AI"     # ❌ Error (different types)


So data type defines:

* What you can do
* What you cannot do

---

## 3️⃣ Memory Management

Different data types use memory differently.

* `int` uses memory for numbers
* `float` uses memory for decimal precision
* `list` stores multiple values
* `dict` stores key-value pairs

Efficient AI systems require proper data type usage for performance.

---

## 4️⃣ Improve Code Clarity

When you use correct data types:

* Code becomes readable
* Bugs reduce
* Debugging becomes easier

Example:


is_trained = True   # clearly boolean

## 5️⃣ Prevent Logical Errors

Example:

```python
age = "25"   # string
age + 5      # ❌ error
```

Correct:


age = 25
age + 5      # ✔ works


# 🧠 Simple Definition 

> Data types define the type of data a variable can store and determine the operations that can be performed on it.

# 🔹 Data Type Categories in Python

Python data types are mainly divided into **6 major categories**:

---

## 1️⃣ Numeric Types

Used to store numbers.

* `int` → Integer (10, -5, 200)
* `float` → Decimal numbers (3.14, -2.5)
* `complex` → Complex numbers (2 + 3j)

Example:

x = 10

y = 3.14

z = 2 + 3j


## 2️⃣ Sequence Types

Ordered collection of elements.

* `str` → String (text)
* `list` → Ordered & mutable collection
* `tuple` → Ordered & immutable collection

Example:

name = "AI"

nums = [1, 2, 3]

t = (1, 2, 3)


## 3️⃣ Set Types

Unordered collection of unique elements.

* `set`
* `frozenset` (immutable set)

Example:

s = {1, 2, 3}


## 4️⃣ Mapping Type

* `dict` → Key-value pairs

Example:

student = {"name": "Phani", "marks": 95}


## 5️⃣ Boolean Type

* `bool` → `True` or `False`

Example:

is_ready = True


## 6️⃣ None Type

* `NoneType` → Represents no value

Example:

x = None

# Base conversion techniques in Python 

In Python, we can convert numbers between:

* **Decimal (base 10)**
* **Binary (base 2)**
* **Octal (base 8)**
* **Hexadecimal (base 16)**

---

# 🔹 1️⃣ Decimal → Binary, Octal, Hex

Python has built-in functions:

* `bin()` → Binary
* `oct()` → Octal
* `hex()` → Hexadecimal

### Example:

n = 25

print(bin(n))   # Decimal to Binary
print(oct(n))   # Decimal to Octal
print(hex(n))   # Decimal to Hex

### Output:

```
0b11001
0o31
0x19
```

👉 `0b` → binary prefix
👉 `0o` → octal prefix
👉 `0x` → hexadecimal prefix

---

# 🔹 2️⃣ Binary → Decimal

## Use `int(value, base)`

b = "11001"
print(int(b, 2))


Output:

```
25


# 🔹 3️⃣ Octal → Decimal

o = "31"
print(int(o, 8))

Output:

25


# 🔹 4️⃣ Hexadecimal → Decimal


h = "19"
print(int(h, 16))

Output:

25


# 🔥 Direct Base Conversions

### Binary → Hex

```python
print(hex(int("11001", 2)))
```

### Hex → Binary

```python
print(bin(int("19", 16)))
```

---

# 🔥 Removing Prefix (Interview Trick)


n = 25

print(bin(n)[2:])  # Remove 0b

Output:

```
11001
```

# Complex data types 

###  Complex numbers are important in mathematics, signal processing, and sometimes in AI-related numerical computations.

# 🔹 Complex Data Type in Python

A **complex number** has:

[a + bj ]

Where:

* `a` → real part
* `b` → imaginary part
* `j` → imaginary unit (√-1)

👉 In Python, we use **`j`**, not `i`.

---

# 🔹 Creating Complex Numbers

x = 3 + 4j

print(x)


print(type(x))


### Output:

(3+4j)

<class 'complex'>

# 🔹 Real and Imaginary Parts


x = 3 + 4j

print(x.real)   # Real part

print(x.imag)   # Imaginary part
```

Output:

```
3.0
4.0
```

# 🔹 Another Way to Create

Using `complex()` function:


x = complex(3, 4)

print(x)
```

Output:

```
(3+4j)
```

# 🔹 Arithmetic Operations


a = 2 + 3j
b = 1 + 4j

print(a + b)   # Addition
print(a - b)   # Subtraction
print(a * b)   # Multiplication
print(a / b)   # Division
```

Python automatically handles complex arithmetic.

---

# 🔹 Important Notes

* Complex numbers cannot be compared using `<`, `>`


print(3+4j > 2+1j)   # ❌ Error

Because complex numbers do not have ordering.

---

# 🔹 Conjugate of Complex Number


x = 3 + 4j

print(x.conjugate())
```

Output:

```
(3-4j)
```

---

# 🔹 Absolute Value (Magnitude)

[|a + bj| = \sqrt{a^2 + b^2}]

x = 3 + 4j

print(abs(x))
```

Output:

```
5.0
```

Because:
[sqrt{3^2 + 4^2} = 5]
