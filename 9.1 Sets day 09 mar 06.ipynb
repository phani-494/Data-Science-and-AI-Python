
### Set Category Data Types


=> The purpose of Set Category Data Types is that **"To store Multiple Values either of Same Type OR Different Type OR Both the 
     Types in single Object with Unique Elements Only (Duplicates are not allowed)"**.

=> We have 2 Data Types in Set Category. They are:

1. **set** ( Mutable and Immutable )
2. **frozenset** ( Immutable )



### **Overview of 'set' Data Type**

* **Definition**: 'set' is one of the Pre-Defined Classes and is treated as a Set Data Type.
* **Purpose**: The purpose of the set data type is to store multiple values of the same type, different types, or both in a single object with **unique elements only** (duplicates are not allowed).
* **Storage**: Values of a set must be stored within **Curly Braces `{ }**` and must be separated by commas.
* **Syntax**: `setobj = {Val1, Val2, ..., Val-n}`
* **Ordering**: A Set Object never maintains **Insertion Order** because the PVM (Python Virtual Machine) displays any of the possibilities of the elements of the set.
* **Operations**: You cannot perform **Indexing and Slicing** operations on a set object because it never maintains insertion order.


# 1. What is a Set in Python?

A **set** is an **unordered collection of unique elements**.

Definition:

> A set is a data structure used to store **multiple values without duplicates**.

Example:

```python
s = {10,20,30,40}
print(s)
```

Output (order may change)

```
{40, 10, 20, 30}
```

---

# 2. Important Properties of Sets

### 1️⃣ Unordered

Sets **do not maintain insertion order**.

Example:

```python
s = {1,2,3,4}
print(s)
```

Output may be

```
{3,1,4,2}
```

Reason: sets use **hash tables** internally.

---

### 2️⃣ No Duplicate Elements

Duplicates are **automatically removed**.

Example:

```python
s = {10,20,10,30,20}

print(s)
```

Output

```
{10,20,30}
```

Python keeps **only unique values**.

---

### 3️⃣ Mutable

Sets **can be modified**.

Example:

```python
s = {1,2,3}

s.add(4)

print(s)
```

Output

```
{1,2,3,4}
```

---

### 4️⃣ Elements Must Be Immutable

Set elements **must be hashable**.

Allowed:

```
int
float
string
tuple
```

Not allowed:

```
list
set
dictionary
```

Example:

❌

```python
s = {[1,2,3]}
```

Error:

```
TypeError: unhashable type: 'list'
```

# 4. Accessing Elements

Sets **do not support indexing**.

❌

```python
s = {10,20,30}

print(s[0])
```

Error

```
TypeError
```

Because sets are **unordered**.

---

### Correct Way

Use a loop.

```python
s = {10,20,30}

for i in s:
    print(i)
```

# 7. Set Operations (Very Important)

These come from **mathematics**.

---

### Union

Combines sets.

```python
a = {1,2,3}
b = {3,4,5}

print(a | b)
```

Output

```
{1,2,3,4,5}
```

or

```python
a.union(b)
```

---

### Intersection

Common elements.

```python
print(a & b)
```

Output

```
{3}
```

or

```python
a.intersection(b)
```

---

### Difference

Elements in first set but not second.

```python
print(a - b)
```

Output

```
{1,2}
```

---

### Symmetric Difference

Elements in **either set but not both**.

```python
print(a ^ b)
```

Output

```
{1,2,4,5}
```

# 9. Frozen Set (Advanced Concept)

A **frozen set** is an **immutable set**.

Example:

```python
fs = frozenset([1,2,3])

print(fs)
```

Output

```
frozenset({1, 2, 3})
```

You **cannot modify it**.

# 10. Set vs List

| Feature   | Set           | List      |
| --------- | ------------- | --------- |
| Order     | ❌ No          | ✅ Yes     |
| Duplicate | ❌ Not allowed | ✅ Allowed |
| Indexing  | ❌ No          | ✅ Yes     |
| Speed     | Faster search | Slower    |

# 11. Internal Working (Important)

Sets use **hash tables** internally.

Meaning:

```
element → hash value → storage location
```

This makes:

```
search → O(1)
```

Very fast.
