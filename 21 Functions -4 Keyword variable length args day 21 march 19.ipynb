

## 5) Key Word Variables Length Parameters (or) arguments:

**=>** When we have family of multiple function calls with Key Word Variable number of values / arguments then with normal python programming, we must define multiple function definitions. This process leads to more development time.

**=>** To overcome this process, we must use the concept of **Keyword Variable length Parameters**.

**=>** To Implement Keyword Variable length Parameters concept, we must define single Function Definition and takes a formal Parameter preceded with a symbol called **double asterisk (** param)** and the formal parameter with double asterisk symbol is called **Keyword Variable length Parameters** and whose purpose is to hold / store any number of **(Key, Value)** coming from similar function calls and whose type is **<class, 'dict'>**.

### Syntax for function definition with Keyword Variables Length Parameters:
-------------------------------------------------------------------------------------------------
```python
def functionname(list of formal params, **param):
    --------------------------------------------
    --------------------------------------------
    --------------------------------------------
```

**=>** Here ****param** is called Keyword Variable Length parameter and it can hold any number of Key word argument values (or) Keyword variable number of argument values and ****param** type is **<class, 'dict'>**.




**=> Rule:-** The ****param** must always be written at the last part of the Function Heading and it must be only one (but not multiple).

### Final Syntax for defining a Function:
-------------------------------------------------------------------------
```python
def funcname(PosFormal parms, *Varlenparam, default params, **kwdvarlenparam):
    --------------------------------------------
    --------------------------------------------
    --------------------------------------------
```
================================== **X** ==================================

---

### Quick Breakdown of the Parameters:

To help you remember the order (which is a common point of confusion), here is what that final syntax represents:

1.  **PosFormal parms:** Standard required arguments (e.g., `a, b`).
2.  **\*Varlenparam:** Often called `*args`; collects extra positional arguments into a **tuple**.
3.  **default params:** Arguments with assigned values (e.g., `c=10`).
4.  **\*\*kwdvarlenparam:** Often called `**kwargs`; collects extra keyword arguments into a **dictionary**.



# 📘 **Keyword Variable-Length Arguments (`**kwargs`)**

---

## 🔹 1. Concept (Simple Understanding)

👉 `**kwargs` allows a function to accept **any number of keyword arguments**

> Keyword arguments = arguments passed using **name = value**

---

## 🔹 2. Why Do We Need It?

Sometimes we don’t know:

* How many inputs user will give ❓
* What fields will come ❓

👉 So we use `**kwargs` to handle **dynamic data**

---

## 🔹 3. Syntax

```python
def function_name(**kwargs):
    print(kwargs)
```

👉 `kwargs` is a **dictionary** internally

---

## 🔹 4. Basic Example

```python
def display(**kwargs):
    print(kwargs)

display(name="Phani", age=22, city="Bangalore")
```

### ✅ Output:

```
{'name': 'Phani', 'age': 22, 'city': 'Bangalore'}
```

👉 Stored as:

* key → parameter name
* value → actual value

---

## 🔹 5. Accessing Values

```python
def display(**kwargs):
    for key, value in kwargs.items():
        print(key, ":", value)

display(name="Phani", age=22)
```

---

## 🔹 6. Real-Time Use Cases (VERY IMPORTANT 🔥)

---

### 🛒 1. E-commerce Product Details

```python
def product_info(**details):
    for key, value in details.items():
        print(f"{key} : {value}")

product_info(name="Laptop", price=50000, brand="HP", warranty="2 years")
```

👉 Flexible because:

* Some products have warranty, some don’t
* Some have color, size, etc.

---

### 🧾 2. User Profile System

```python
def user_profile(**data):
    print(data)

user_profile(name="Rahul", email="rahul@gmail.com", phone=9876543210)
```

👉 Used in:

* Registration forms
* APIs

---

### 🌐 3. API Data Handling (Advanced Use)

```python
def send_data(**payload):
    print("Sending:", payload)

send_data(id=101, status="success", amount=500)
```

---

## 🔹 7. Mixing with Other Arguments

```python
def func(a, b, **kwargs):
    print(a, b)
    print(kwargs)

func(10, 20, x=100, y=200)
```

---

## 🔹 8. Important Rules

✅ Must come **last** in parameters

```python
def func(a, **kwargs):  # correct
```

❌ Wrong:

```python
def func(**kwargs, a):  # error
```

---

## 🔹 9. Difference: `*args` vs `**kwargs`

| Feature | `*args`    | `**kwargs`   |
| ------- | ---------- | ------------ |
| Type    | Tuple      | Dictionary   |
| Input   | Positional | Keyword      |
| Example | (1,2,3)    | name="Phani" |

---

## 🔹 10. Common Mistakes ❌

### ❌ Forgetting keyword format

```python
func(10, 20)  # won't go into kwargs
```

### ❌ Treating kwargs like list

```python
kwargs[0]  # wrong
```

---
