# 🔹 What is `range()` in Python?

> `range()` is a built-in function used to generate a sequence of numbers.

It is mostly used in:

* `for` loops
* Iterations
* Generating number sequences efficiently

---

# 🔹 Syntax

range(start, stop, step)
```

### Parameters:

| Parameter | Meaning                                  |
| --------- | ---------------------------------------- |
| start     | Starting value (default = 0)             |
| stop      | Ending value (NOT included)              |
| step      | Difference between numbers (default = 1) |

---

# 🔥 Important Rule

> `stop` value is **excluded**

---

# 🔹 1️⃣ Only Stop Value

range(5)
```

Equivalent to:

range(0, 5, 1)

Example:

for i in range(5):
    print(i)
```

Output:

```
0
1
2
3
4
```

---

# 🔹 2️⃣ Start and Stop

for i in range(2, 6):
    print(i)
Output:

```
2
3
4
5
```

---

# 🔹 3️⃣ Start, Stop, Step

for i in range(1, 10, 2):
    print(i)
    
Output:

```
1
3
5
7
9
```

---

# 🔹 Negative Step (Very Important)

for i in range(10, 0, -2):
    print(i)
    
Output:

```
10
8
6
4
2
```

⚠️ Rule:

If step is negative:

* start must be greater than stop

---

# 🔹 What Type Does `range()` Return?

r = range(5)
print(type(r))

Output:

```
<class 'range'>
```

It does NOT create a list.

It creates a **range object** (memory efficient).

---

# 🔥 Why Is Range Memory Efficient?

If you do:

```python
range(1000000000)
```

It does NOT store 1 billion numbers.

It only stores:

* start
* stop
* step

Numbers are generated **on demand**.

This is why it is very powerful.

---

# 🔹 Convert Range to List

list(range(5))
```

Output:

[0, 1, 2, 3, 4]
```

Now memory is used.

---

# 🔹 Range Object Properties

```python
r = range(1, 10, 2)

print(r.start)
print(r.stop)
print(r.step)
```
# 🔹 Range Supports Indexing

r = range(10)
print(r[3])


Output:

3


##  Range Supports Slicing

r = range(10)
print(r[2:8:2])

Output:

range(2, 8, 2)

Notice:
It returns another range object — not list.

---

# 🔥 Very Important Concept

`range()` is:

* Immutable
* Iterable
* Memory efficient
* Lazy sequence generator

---

#  Edge Cases

### Step cannot be 0


range(1, 10, 0)  # ❌ Error
```

Error:

ValueError: range() arg 3 must not be zero


---

# 🔥 Internal Logic Formula

For positive step:

start < stop
```

For negative step:

start > stop

What is the output?

```python
list(range(5, 5))
```

Answer:

```
[]
```

Because start == stop → nothing generated.

---

# 🔹 One-Line Definition for Notes

> `range()` generates an immutable sequence of numbers lazily using start, stop, and step values.


````python
for i in range(10,10): # nothing generated since start == stop 
    print(i)
````

````python
for i in range(5,51,5): # gives 5 table 
    print(i)
````

````python
for i in range(-9,0,1):
    print(i)
````

````python
a=range(10)
a[2]=100
````

## Lists 

## 'list'** is one of the pre-defined classes and is treated as a **List Category Data Type**.

The purpose of the list Data Type is to **"Store Multiple Values either of Same Type OR Different Type OR Both the
Types in a single Object with Unique and Duplicates"**.

To store the List data, we use **Square Brackets**. In other words, the elements of a list must be enclosed within Square 
Brackets `[ ]` and values must be separated by commas.

> **Syntax:** `varname = [Val1, Val2, ......, Val-n]`
> *Here, **varname** is an object of list.*

An object of list **Maintains Insertion Order**.

- Since lists maintain insertion order, they are also **indexable**.
- on the object of list , we can perform indexing and slicing operations 

````python
l1=[100,200,200,45,-54]
l1
````

````python
l2=["indra", " ishq",45.6,False,4+5j]
print(l2)
````

````python
print(l2[0])
print(l2[-len(l2)])
print(l2[len(l2)-1])
print(l2[-1])
````

````python
l2[100]
````

````python
l2[1:5]
````

````python
l2[::2]
````

````python
l2[::-1]
````

````python
print(l2,type(l2),id(l2))
````

````python
l2[1]='SIVA'
print(l2)
````

````python
print(l2,type(l2),id(l2)) # modification done in same memory location only. so mutable 
````

````python
l3=["india", " ishq",45.6,False,4+5j]
l4=["india", " ishq",45.6,False,4+5j]
print(l3==l4)
print(id(l3),id(l4)) # both are equal but stored in the different memory locations 
````

````python
print(l3)
# slicing based assignment 
l3[1:3]=["tqu",90] # values at indexes 1 and 2 are changing here 
print(l3)
````

````python
print(l4)
l4[::2]=["maha",89,67.90] # updating elements at the step position values 
print(l4)
````

# Empty and non-empty lists
* An object of list belongs to **MUTABLE** because list objects allow us to perform Modification / Updations / Assignments by 
using Indexing and Slicing.
* We can create **Two types** of list objects. They are:
1. Empty list
2. Non-Empty list


### **Empty List**

An empty list is one which does not contain any elements and whose length is zero.
    
* **Syntax1:** `listobj=[]`
* **Syntax2:** `listobj=list()`

### **Non-Empty List**

A Non-Empty list is one which contains Elements and whose length is > 0.

* **Syntax1:** `listobj=[Val1,Val2,....,Val-n]`


````python
name="phanindra reddy"
s1=list(name)
print(s1)
s2=set(s1)
print(s2)
````

## **Pre-Defined Functions in list**

* We know that on the object of list, we can perform both Indexing and Slicing Operations.
* With Indexing, we get / Access Single Value and we can update single value of List object.
* With Slicing Operation we get / access multiple values and we can update Multiple Values of list object.
* Hence with index and slicing Operation on list, we can access and update value(s).
* To do More Operation on list object along with index and slicing Operation, we use the pre-defined functions present in list object.


1. append()
-  used to add the elements/values at the end of the list (called appending)

````python
l1=[100,200,200,45,-54]
print(l1,id(l1))
````

````python
l1.append("Python")
print(l1,id(l1))
````

````python
l1.append(899)
````

````python
print(l1)
````
