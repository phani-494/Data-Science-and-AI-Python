## `print()` in Python

`print()` is a **built-in function** used to **display output on the screen (console)**.

It is the **most commonly used function** in Python.

---

# 1️⃣ Basic Syntax

```python
print(object1, object2, ...)
```

It prints the **values passed inside it**.

---

# 2️⃣ Simple Example

```python
print("Hello Python")
```

Output

```
Hello Python
```

---

# 3️⃣ Printing Variables

```python
name = "Phanindra"
age = 22

print(name)
print(age)
```

Output

```
Phanindra
22
```

---

# 4️⃣ Printing Multiple Values

```python
print("Python", "AI", "ML")
```

Output

```
Python AI ML
```

By default Python **separates values with a space**.

---

# 5️⃣ `sep` Parameter (Separator)

`sep` changes the **separator between values**.

Example

```python
print("Python", "AI", "ML", sep="-")
```

Output

```
Python-AI-ML
```

Another example

```python
print(10,20,30, sep="*")
```

Output

```
10*20*30
```

---

# 6️⃣ `end` Parameter

By default `print()` ends with a **newline (`\n`)**.

Example

```python
print("Hello")
print("World")
```

Output

```
Hello
World
```

Using `end`:

```python
print("Hello", end=" ")
print("World")
```

Output

```
Hello World
```

---

# 7️⃣ Printing Expressions

```python

```

Output

```
8
```

## `input()` in Python

`input()` is a **built-in function used to take input from the user through the keyboard**.

When Python executes `input()`, the program **waits until the user enters something**.

---

# 1️⃣ Basic Syntax

```python
input()
```

or

```python
input("message")
```

The message is called a **prompt**.

---

# 2️⃣ Simple Example

```python
name = input("Enter your name: ")
print(name)
```

Example run:

```
Enter your name: Phanindra
Phanindra
```

Here:

* User types **Phanindra**
* Python stores it in variable **name**

---

# 3️⃣ Important Rule

`input()` **always returns a string**.

Example:

```python
x = input("Enter a number: ")
print(type(x))
```

Input:

```
10
```

Output:

```
<class 'str'>
```

Even though you typed **10**, Python treats it as **"10" (string)**.

---

# 4️⃣ Converting Input to Integer

To perform calculations, convert it using **typecasting**.

### Integer input

```python
x = int(input("Enter a number: "))
print(x + 5)
```

Example:

```
Enter a number: 10
15
```

---

### Float input

```python
x = float(input("Enter a decimal number: "))
print(x)
```

Example:

```
Enter a decimal number: 3.5
3.5
```

---

# 5️⃣ Multiple Inputs

### Method 1 (separate inputs)

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print(a + b)
```

---

### Method 2 (single line input)

```python
a, b = map(int, input("Enter two numbers: ").split())
print(a + b)
```

Example input:

```
10 20
```

Output:

```
30
```

---

# 6️⃣ Example Program

```

Example run:

```
