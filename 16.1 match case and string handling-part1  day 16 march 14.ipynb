# match case 

## `match-case` in Python (Pattern Matching)

`match-case` is a **control flow statement** introduced in **Python 3.10**.
It works similar to **switch-case in other languages**, but it is **much more powerful** because it supports **pattern matching**.

It compares a value against **different patterns** and executes the matching block.

---

# 1️⃣ Basic Syntax

```python
match variable:
    case value1:
        # code
    case value2:
        # code
    case _:
        # default case
```

`_` means **default case** (like `else`).

---

# 2️⃣ Simple Example

```python
num = int(input("Enter number: "))

match num:
    case 1:
        print("One")
    case 2:
        print("Two")
    case 3:
        print("Three")
    case _:
        print("Invalid number")
```

### Example Output

```
Input: 2
Output: Two
```

---

# 4️⃣ Multiple Values in One Case

```python
num = int(input())

match num:
    case 1 | 2 | 3:
        print("Small number")
    case 4 | 5 | 6:
        print("Medium number")
    case _:
        print("Large number")
```

`|` means **OR pattern**.

# Why we use `|` in `match-case`

In **Python**, the symbol `|` inside a `case` means **OR pattern**.

It tells Python:

👉 **Match if the value is ANY ONE of the given options.**

So Python checks multiple possibilities in a **single case block**.

---

# Example from your code

```python
num = int(input())

match num:
    case 1 | 2 | 3:
        print("Small number")
```

### What Python checks internally

It behaves like:

```python
if num == 1 or num == 2 or num == 3:
    print("Small number")
```

So if the input is:

```
1
2
3
```

Output:

```
Small number
```

## Python `match case` Statement

* **Introduction:** The `match case` statement is one of the conditional statements available from **Python 3.10 version** onwards.
* **Purpose:** The purpose of the `match case` statement is "To Deal with Pre-Designed Conditions OR Menu Driven Applications".
* **Menu Driven Applications:** These applications contain pre-designed conditions.

---

### Syntax:

```python
match(Choice Expr):
    case Choice Label1:
        Block of Statements-1
    case Choice Label2:
        Block of Statements-2
    case Choice Label3:
        Block of Statements-3
    -------------------------
    case Choice Label-n:
        Block of Statements-n
    case _:  # Default Case Block
        default Block of Statements
---------------------------------------------
Other Statements in Program

```

---

### Key Notes:

* **Pattern Matching:** The `match` keyword takes an expression and compares its value to successive patterns given as `case` blocks.
* **The Wildcard (`_`):** The `case _:` acts as a wildcard or "default" case. It will execute if none of the specific labels above it match the choice expression.
* **Structural Comparison:** Unlike traditional `if-elif-else` chains, this is designed for structural pattern matching, making menu-driven logic much cleaner.




## Explanation:

* **Keywords:** Here, `match` and `case` are the keywords.(proposed keywords not the standard keywords)
* **Choice Expr:** The "Choice Expr" represents either **int**, **str**, or **bool**.
* **Execution Flow:**
* If **"Choice Expr"** matches **"case label1"**, then the PVM executes **Block of Statements-1** and later executes **Other statements** in the program.
* If **"Choice Expr"** matches **"case label2"**, then the PVM executes **Block of Statements-2** and later executes **Other statements** in the program.
* In general, the **"Choice Expr"** tries to match with case label-1, case label-2, ..., case label-n. The PVM then executes the corresponding block of statements and later executes **Other statements** in the program.


* **Default Case:** If **"Choice Expr"** does not match any of the specified case labels, then the PVM executes the **Default Block of Statements** written under the default case block (`case _`) and later executes **Other statements** in the program.
* **Optionality:** Writing the default case block is **optional**. However, if you do write it, it **must be written last** (otherwise, you will get a `SyntaxError`).
* **Multiple Labels:** When representing multiple case labels under one case, those case labels must be combined with the **Bitwise OR Operator (`|`)**.

---

### Quick Tip on the `|` (Pipe) Operator:

If you want one block of code to run for multiple choices (like "Yes" or "Y"), you would write it like this:

```python
case "Yes" | "Y":
    print("User agreed!")

```


# Temperature conversion calculator 

````python

print("--- Temperature Conversion Calculator ---")
print("1. F to C (Fahrenheit to Celsius)")
print("2. F to K (Fahrenheit to Kelvin)")
print("3. C to F (Celsius to Fahrenheit)")
print("4. C to K (Celsius to Kelvin)")
print("5. K to F (Kelvin to Fahrenheit)")
print("6. K to C (Kelvin to Celsius)")
print("7. Exit")

choice = input("Enter Ur Choice: ")

match choice:
    case "1":
        f = float(input("Enter Temp in F: "))
        print(f"Celsius: {(f - 32) * 5/9:.2f}")
    case "2":
        f = float(input("Enter Temp in F: "))
        print(f"Kelvin: {(f - 32) * 5/9 + 273.15:.2f}")
    case "3":
        c = float(input("Enter Temp in C: "))
        print(f"Fahrenheit: {(c * 9/5) + 32:.2f}")
    case "4":
        c = float(input("Enter Temp in C: "))
        print(f"Kelvin: {c + 273.15:.2f}")
    case "5":
        k = float(input("Enter Temp in K: "))
        print(f"Fahrenheit: {(k - 273.15) * 9/5 + 32:.2f}")
    case "6":
        k = float(input("Enter Temp in K: "))
        print(f"Celsius: {k - 273.15:.2f}")
    case "7":
        print("Exiting... Goodbye!")
    case _:
        print("Invalid Choice! Please select 1-7.")

print("-" * 40)
print("Process Completed.")
````

---

### How it works:

* **`match choice:`**: It takes your input and looks for the corresponding `case`.
* **`case _:`**: This is the "Default Case" from your notes. If the user types "9" or "Hello", it catches the error and prints "Invalid Choice".
* **`.2f`**: This is just a little extra formatting to make sure the result only shows 2 decimal places (like **25.56** instead of **25.555555556**).
