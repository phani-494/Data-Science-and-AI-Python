## Python Decorators 

 A decorator is essentially a function that takes another function, adds some "bonus" behavior to it, and returns a new version of it.

To help you move from the theory in that screenshot to actual code, let's break down how this works with a concrete example.

---

## Python Decorator Structure

Think of a decorator like a **gift wrap**. The gift is your "Normal Function," and the wrapping paper is the "Decorator" that makes the gift look better or adds a card to it without changing what’s inside the box.


### 1. The Real-World Syntax
, here is how you would write a decorator that adds a simple "Hello" and "Goodbye" message to any function:

```python
# The Decorator (Outer Function)
def my_decorator(func):
    # The Wrapper (Inner Function)
    def wrapper():
        print("Something is happening before the function is called.")
        func() # This is the 'Normal Function' being executed
        print("Something is happening after the function is called.")
    
    return wrapper

# Using the decorator with the @ symbol
@my_decorator
def say_hello():
    print("Hello!")

say_hello()
```

### 2. Key Takeaways :
* **Nested Functions:** The `innerfunction` (often called a `wrapper`) is where the "magic" happens. It lives inside the `OuterFunctionName1`.
* **Returning the Function:** `return innerfunction`. This is crucial. You aren't returning the *result* of the inner function; you are returning the **function itself** so it can be called later.
* **Modification:**This is commonly used for logging, timing how long a function takes, or checking if a user is logged in before letting them run a specific task.

---

### Why use this?
Instead of rewriting the same code (like logging or security checks) inside 50 different functions, you write it **once** as a decorator and just "tag" your other functions with it. It keeps your code DRY (**D**on't **R**epeat **Y**ourself).

````python

### How to use the "Decorator" 
#Instead of calling `n=getval()` inside every single function, you can use a decorator to "inject" that value automatically. 


def get_value_decorator(func):
    def innerfunction():
        n = getval() # Get the result of Normal Function (getval)
        return func(n) # Apply modification (square, cube, etc.)
    return innerfunction

@get_value_decorator
def square(n):
    res = n**2
    print("Square({})={}".format(n, res))

@get_value_decorator
def cube(n):
    res = n**3
    print("Cube({})={}".format(n, res))

# Now when you call square(), the decorator automatically gives it '5'
square() 
````

Think of a Python decorator as **gift wrapping** for a function. The gift (your function) stays exactly the same inside, but the wrapping (the decorator) can change how it looks or add extra flair before it’s even opened.

In technical terms, a decorator is a function that takes another function as an argument, adds some functionality, and returns a new function—all without permanently modifying the original source code.

---

## 1. The Core Prerequisites
To understand decorators, you need to accept two "Pythonic" truths:
1.  **Functions are first-class objects:** You can assign them to variables, pass them as arguments, and return them from other functions.
2.  **Closures:** Inner functions can "remember" the variables defined in their local scope even after the outer function has finished executing.

---

## 2. The Anatomy of a Decorator
Here is how a decorator is constructed from scratch. We’ll build one that simply logs when a function is being called.

### The Manual Way
```python
def my_logger(func):
    def wrapper():
        print(f"Logging: Running {func.__name__}...")
        func()
        print("Logging: Finished!")
    return wrapper

def say_hello():
    print("Hello, world!")

# Decorating manually
decorated_hello = my_logger(say_hello)
decorated_hello()
```

### The "Sugar" Way (`@`)
Python provides the `@` symbol (syntactic sugar) to make this look much cleaner. It does exactly what we did above behind the scenes.

```python
@my_logger
def say_hello():
    print("Hello, world!")

say_hello()
```



---

## 3. Handling Arguments: `*args` and `**kwargs`
The example above works for functions with no arguments. To make a decorator truly universal, the `wrapper` function must accept any number of positional and keyword arguments.

```python
def universal_decorator(func):
    def wrapper(*args, **kwargs):
        print("Something is happening before the function.")
        result = func(*args, **kwargs)  # Execute the original function
        print("Something is happening after the function.")
        return result  # Return the result so the caller gets it
    return wrapper
```

---

## 4. Why Use Them? (Real-World Use Cases)
Decorators are perfect for "cross-cutting concerns"—tasks that you want to apply to many different functions without repeating code.

| Use Case | Description |
| :--- | :--- |
| **Logging** | Tracking when functions are called and with what data. |
| **Timing** | Measuring how long a function takes to run (performance profiling). |
| **Authorization** | Checking if a user has permission to run a specific web route (common in Flask/Django). |
| **Caching** | Saving the results of expensive function calls to return them faster next time. |

---

## 5. A Deep Dive: Keeping the Metadata
When you decorate a function, it technically becomes the `wrapper` function. This means its name and docstring are "lost."

```python
print(say_hello.__name__) # This would output 'wrapper', not 'say_hello'
```

To fix this, we use `functools.wraps`. This is a decorator for your decorator! It copies the original function’s identity to the wrapper.

```python
from functools import wraps

def smart_decorator(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        return func(*args, **kwargs)
    return wrapper
```

---

## 6. Decorators with Arguments
Sometimes you want to pass configuration into the decorator itself, like `@repeat(times=3)`. This requires **three** layers of functions:
1.  The outermost function to handle the arguments.
2.  The middle function to act as the decorator.
3.  The innermost function to act as the wrapper.



First, let's see how a decorator actually works — it's a function that wraps another function:

The `@` syntax is just shorthand — Python replaces your function with a wrapped version that adds extra behaviour before and/or after the original runs.

Now let's go through every concept with full examples:Here's a quick summary of everything covered:

**Core idea** — a decorator is just a function that takes a function and returns a new, enhanced function. The `@` syntax is pure shorthand.

**The 4 rules to never forget:**
1. Always use `@wraps(func)` inside your decorator — it preserves the original function's name and docstring
2. Always accept `*args, **kwargs` in your wrapper so it works with any function signature
3. Always `return` the wrapper function at the end of the decorator
4. Decorators stack bottom-up but execute top-down

# ⚡ One Line to Remember Forever

---



Think of a Python decorator as **gift wrapping** for a function. The gift (your function) stays exactly the same inside, but the wrapping (the decorator) can change how it looks or add extra flair before it’s even opened.

In technical terms, a decorator is a function that takes another function as an argument, adds some functionality, and returns a new function—all without permanently modifying the original source code.

---

## 1. The Core Prerequisites
To understand decorators, you need to accept two "Pythonic" truths:
1.  **Functions are first-class objects:** You can assign them to variables, pass them as arguments, and return them from other functions.
2.  **Closures:** Inner functions can "remember" the variables defined in their local scope even after the outer function has finished executing.

---

## 2. The Anatomy of a Decorator
Here is how a decorator is constructed from scratch. We’ll build one that simply logs when a function is being called.

### The Manual Way
```python
def my_logger(func):
    def wrapper():
        print(f"Logging: Running {func.__name__}...")
        func()
        print("Logging: Finished!")
    return wrapper

def say_hello():
    print("Hello, world!")

# Decorating manually
decorated_hello = my_logger(say_hello)
decorated_hello()
```

### The "Sugar" Way (`@`)
Python provides the `@` symbol (syntactic sugar) to make this look much cleaner. It does exactly what we did above behind the scenes.

```python
@my_logger
def say_hello():
    print("Hello, world!")

say_hello()
```



---

## 3. Handling Arguments: `*args` and `**kwargs`
The example above works for functions with no arguments. To make a decorator truly universal, the `wrapper` function must accept any number of positional and keyword arguments.

```python
def universal_decorator(func):
    def wrapper(*args, **kwargs):
        print("Something is happening before the function.")
        result = func(*args, **kwargs)  # Execute the original function
        print("Something is happening after the function.")
        return result  # Return the result so the caller gets it
    return wrapper
```

---

## 4. Why Use Them? (Real-World Use Cases)
Decorators are perfect for "cross-cutting concerns"—tasks that you want to apply to many different functions without repeating code.

| Use Case | Description |
| :--- | :--- |
| **Logging** | Tracking when functions are called and with what data. |
| **Timing** | Measuring how long a function takes to run (performance profiling). |
| **Authorization** | Checking if a user has permission to run a specific web route (common in Flask/Django). |
| **Caching** | Saving the results of expensive function calls to return them faster next time. |

---

## 5. A Deep Dive: Keeping the Metadata
When you decorate a function, it technically becomes the `wrapper` function. This means its name and docstring are "lost."

```python
print(say_hello.__name__) # This would output 'wrapper', not 'say_hello'
```

To fix this, we use `functools.wraps`. This is a decorator for your decorator! It copies the original function’s identity to the wrapper.

```python
from functools import wraps

def smart_decorator(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        return func(*args, **kwargs)
    return wrapper
```

## 6. Decorators with Arguments
Sometimes you want to pass configuration into the decorator itself, like `@repeat(times=3)`. This requires **three** layers of functions:
1.  The outermost function to handle the arguments.
2.  The middle function to act as the decorator.
3.  The innermost function to act as the wrapper.
