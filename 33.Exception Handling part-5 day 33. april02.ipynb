# generate multiplication table for a given number
- if number is -ve, then raise NegNumError
- if number is 0, then raise ZeroNumError
- if number is alphanumeric , then generate AlphaError
-if number is +ve, then result should be it's table

````python
%%writefile MultiTableExcept.py
class NegNumError(Exception):
    pass
class ZeroNumError(Exception):
    pass
class AlphaNumError(Exception):
    pass
    # if empty input is given we can also give EmptyInputError
````

````python
%%writefile MulTableOperation1.py
from MultiTableExcept import *
def multi(num):
    if num<0:
        raise NegNumError
    elif num==0:
        raise ZeroNumError
    elif type(num)!=int or type(num)==str:
        raise AlphaNumError
    else:
        return num
````

````python
from MultiTableExcept import *
from MulTableOperation1 import *
try:
    a=int(input("Enter a number:"))
    res=multi(a)
except ValueError: # already predefined
    print("You have entered a string")
except NegNumError:
    print("You have entered a Negative Number: Try again")
except ZeroNumError:
    print("You entered 0- an invalid number")
except AlphaNumError:
    print("You haven't entered a integer")
else:
    for i in range(1,11):
        print("{}*{}={}".format(res,i,res*i))
````

# 🏦 ATM System with Exception Handling (Best Example)

```python
# ------------------ exceptions.py ------------------

class InsufficientBalanceError(Exception):
    pass

class InvalidAmountError(Exception):
    pass

class DailyLimitExceededError(Exception):
    pass


# ------------------ atm.py ------------------

from exceptions import InsufficientBalanceError, InvalidAmountError, DailyLimitExceededError

class ATM:
    def __init__(self, balance=0):
        self.balance = balance
        self.daily_limit = 20000

    def deposit(self, amount):
        if amount <= 0:
            raise InvalidAmountError("Deposit amount must be positive")
        
        self.balance += amount
        return self.balance

    def withdraw(self, amount):
        if amount <= 0:
            raise InvalidAmountError("Withdrawal amount must be positive")
        
        if amount > self.daily_limit:
            raise DailyLimitExceededError("Daily withdrawal limit exceeded")
        
        if amount > self.balance:
            raise InsufficientBalanceError("Insufficient balance")
        
        self.balance -= amount
        return self.balance

    def check_balance(self):
        return self.balance


# ------------------ main.py ------------------

from atm import ATM
from exceptions import InsufficientBalanceError, InvalidAmountError, DailyLimitExceededError

atm = ATM(10000)

while True:
    print("\n1. Deposit\n2. Withdraw\n3. Balance\n4. Exit")
    
    try:
        choice = int(input("Enter choice: "))

        if choice == 1:
            amt = int(input("Enter deposit amount: "))
            print("Updated Balance:", atm.deposit(amt))

        elif choice == 2:
            amt = int(input("Enter withdrawal amount: "))
            print("Updated Balance:", atm.withdraw(amt))

        elif choice == 3:
            print("Current Balance:", atm.check_balance())

        elif choice == 4:
            print("Thank you for using ATM")
            break

        else:
            print("Invalid choice")

    except InvalidAmountError as e:
        print("❌ Invalid Amount:", e)

    except InsufficientBalanceError as e:
        print("❌ Transaction Failed:", e)

    except DailyLimitExceededError as e:
        print("❌ Limit Error:", e)

    except ValueError:
        print("❌ Please enter valid numbers only")

    finally:
        print("➡️ Transaction processed")
```

# 🏧 4. `atm.py`

```python id="n2vxz3"
from exceptions import *
from storage import log_transaction

class ATM:
    def __init__(self, account):
        self.account = account

    def authenticate(self, pin):
        if pin != self.account.pin:
            raise AuthenticationError("Invalid PIN")

    def deposit(self, amount):
        if amount <= 0:
            raise InvalidAmountError("Deposit must be positive")

        self.account.balance += amount
        log_transaction(f"Deposited: {amount}")
        return self.account.balance

    def withdraw(self, amount):
        if amount <= 0:
            raise InvalidAmountError("Withdrawal must be positive")

        if amount > self.account.balance:
            raise InsufficientBalanceError("Insufficient balance")

        self.account.balance -= amount
        log_transaction(f"Withdrawn: {amount}")
        return self.account.balance

    def check_balance(self):
        return self.account.balance
```

# 🧠 1. What is an Exception?

👉 An **exception** is:

> An error that occurs during program execution and disrupts normal flow.

# 🧱 5. Full Exception Handling Structure

```python
try:
    pass

    pass

    pass

    pass
```

---

## 🔍 Meaning of each block

### 🔹 `try`

* Contains risky code

---

### 🔹 `except`

* Runs if error occurs

---

### 🔹 `else`

* Runs if NO exception

---

### 🔹 `finally`

* Always runs (important!)

---

## ✅ Example

```python
try:
    x = int(input("Enter: "))
    y = int(input("Enter: "))
    print(x / y)

except ZeroDivisionError:
    print("Cannot divide by zero")

except ValueError:
    print("Invalid input")

else:
    print("Success")

finally:
    print("Done")
```

# 🔄 10. Internal Flow (VERY IMPORTANT)

👉 Step-by-step:

1. Python executes `try`
2. Error occurs
3. Python creates **exception object**
4. Jumps to matching `except`
5. Executes handling code

# 📦 11. Exception Hierarchy (Important Concept)

👉 All exceptions come from:

```

    └── Exception
          ├── ValueError
          ├── TypeError
          ├── IndexError
```

👉 Your custom exceptions must inherit from `Exception`
