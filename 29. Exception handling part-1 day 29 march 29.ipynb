# 🧠 Purpose of Exception Handling in Python

---

## ✅ 1. Definition

> **Exception handling is a mechanism to handle runtime errors so that the program does not crash and continues execution smoothly.**

---

## 🎯 2. Main Purpose (Core Idea)

👉 Without exception handling → program **crashes**
👉 With exception handling → program **handles error gracefully**

---

## 💥 3. Why Do We Need It?

### ❌ Without Exception Handling

```python
a = int(input("Enter number: "))
b = 10 / a
print(b)
```

If user enters `0`:

👉 Program crashes with:

```
ZeroDivisionError
```

---

### ✅ With Exception Handling

```python
try:
    a = int(input("Enter number: "))
    b = 10 / a
    print(b)
except ZeroDivisionError:
    print("Cannot divide by zero")
```

👉 Program does NOT crash
👉 It handles the error safely

---

## ✅ 1. Definition

> A **compile-time error** is an error detected by Python **before execution**, while converting source code into bytecode.

👉 Python stops immediately
👉 Program **does NOT run at all**

---

## 🧠 2. Internal Working (Important for Deep Understanding)

Python execution has 2 main phases:

1. **Compilation Phase**

   * Source code → Bytecode
   * Syntax is checked

2. **Execution Phase**

   * Bytecode → Executed by PVM (Python Virtual Machine)

👉 Compile-time errors occur in **Phase 1**

---

## ⚠️ 3. Key Insight

Even though Python is interpreted:

> It still performs a **compilation step**, so compile-time errors DO exist.

---

## 🚨 4. Types of Compile-Time Errors

### 🔴 1. SyntaxError

> Violating Python grammar rules

#### Example:

```python
if True
    print("Hello")
```

❌ Error:

```
SyntaxError: expected ':'
```

---

### 🔴 2. IndentationError

> Improper indentation (Python is indentation-sensitive)

#### Example:

```python
def fun():
print("Hello")
```

❌ Error:

```
IndentationError: expected an indented block
```

---

### 🔴 3. TabError

> Mixing tabs and spaces

#### Example:

```python
def fun():
	print("Hi")
    print("Hello")
```

❌ Error:

```
TabError: inconsistent use of tabs and spaces
```

---

## 🧠 5. Common Causes

* Missing `:` after `if`, `for`, `while`, `def`, `class`
* Incorrect indentation
* Unmatched brackets `() [] {}`
* Wrong keyword usage
* Invalid syntax structure

---

* ❌ Program does NOT start
* 📍 Error shows exact line number
* ⚡ Easy to detect and fix
* 🧾 Error message is clear

## Building Points in Exception Handling
============================================================

1. **When the Application User Enters Wrong OR Invalid Input then we get Runtime Error.**
   (Invalid Input ------> Runtime Error)

2. **All Runtime Errors by default gives Technical Error Messages.**
   (Invalid Input ------> Runtime Error ------> Technical Error Messages)

3. **Definition of Exception: Every Runtime Error is called Exception**
   (Invalid Input ------> Runtime Error ------> Exception)

   
   **All invalid Inputs are considered as Exceptions**

5. **All Exceptions by default gives Technical Error Messages.**

6. **Definition of Exception Handling:- The Process of Converting Technical Error Messages into User-Friendly Messages is called Exception Handling**




**6. When the exception occurs in Python Program, The Steps Takes Place Internally:**
* i) Program Execution Abnormally Terminated
* ii) PVM Comes out-off Program flow
* iii) PVM by default generates Technical Error Message.

**7. To do the above Step-(i), Step-(ii) and Step-(iii), PVM CREATES AN OBJECT w.r.t An EXCEPTION CLASS.**

**8. When an Exception occurs in Python Program then PVM creates an Object of appropriate EXCEPTION CLASS.**

**9. Hence All Exceptions are Considered as objects of EXCEPTION CLASSES.**

### Key Terms Defined:
* **PVM:** Python Virtual Machine (the engine that runs your Python code).
* **Exception Class:** A blueprint in Python (like `ValueError` or `ZeroDivisionError`) used to create an error object when something goes wrong.


### **1. The Nature of Exceptions**
An **Exception** is essentially a **Runtime Error** that occurs when a user provides invalid or unexpected input while the program is running. Unlike syntax errors (which prevent the code from starting), exceptions happen during execution and disrupt the normal flow.

### **2. The Internal Process (The "3 Abnormals")**
When an error occurs, the **Python Virtual Machine (PVM)** handles it internally through three immediate steps:
1.  **Abnormal Termination:** The program stops running immediately.
2.  **Flow Disruption:** The PVM jumps out of the regular code sequence.
3.  **Technical Feedback:** The PVM generates a "Technical Error Message" (a Traceback), which is often difficult for non-developers to understand.



### **3. The Role of Objects and Classes**
In Python, every exception is treated as an **Object**. 
* When an error occurs, the PVM identifies the type of error (e.g., dividing by zero).
* It then creates an **instance (object)** of a specific **Exception Class** (like `ZeroDivisionError` or `ValueError`) to represent that specific failure.

### **4. The Goal of Exception Handling**
The primary purpose of "Handling" an exception is **Translation**. It is the process of intercepting those scary, technical system messages and converting them into **User-Friendly Messages**. This ensures that the end-user understands what went wrong without seeing the raw code or the program crashing unexpectedly.

---

## ✅ 1. Core Idea

> **`IndexError` is a child class (subclass) of the base class `Exception`.**

That means:
👉 It is a **type of exception**
👉 It belongs to Python’s **exception hierarchy**

---

## 🧬 2. Hierarchy (Very Important)

Python organizes errors like a tree:

```
BaseException
     ↓
  Exception
     ↓
 IndexError
```

👉 So:

* `BaseException` → top level
* `Exception` → general errors
* `IndexError` → specific error

---

## 🧠 3. What This Means Practically

### 🔹 Case 1: Catch Specific Error

```python
try:
    lst = [1, 2, 3]
    print(lst[5])
except IndexError:
    print("Handled IndexError")
```

---

### 🔹 Case 2: Catch General Exception

```python
try:
    lst = [1, 2, 3]
    print(lst[5])
except Exception:
    print("Handled by Exception class")
```

👉 Works because:
**`IndexError` IS-A `Exception`**

---

## ⚡ 4. Key Concept (Inheritance)

> `IndexError` inherits from `Exception`

So:

* It gets all properties of `Exception`
* But represents a **specific situation (index out of range)**

---

# 🧠 What Exactly Happens Internally When an Exception Arises?

---

## ✅ 1. Core Idea

> When an error occurs, Python **creates an exception object and raises it**, interrupting normal program flow.

---

## ⚙️ 2. Step-by-Step Internal Process

Let’s take this example:

```python
a = 10
b = 0
print(a / b)
```

---

### 🔹 Step 1: Code Execution Starts

Python executes line by line using the **Python Virtual Machine (PVM)**

---

### 🔹 Step 2: Problem Occurs

```python
a / b   # 10 / 0
```

👉 Division by zero is **invalid**

---

### 🔹 Step 3: Exception Object is Created

Python internally does something like:

```python
ZeroDivisionError("division by zero")
```

👉 This is an **object** (instance of exception class)

---

### 🔹 Step 4: Exception is “Raised”

> Python **raises** the exception (throws it into the system)

👉 Normal execution **stops immediately**

---

### 🔹 Step 5: Python Searches for Handler

Python looks for a matching `except` block:

```python
try:
    print(a / b)
except ZeroDivisionError:
    print("Handled")
```

👉 If found → handled ✅
👉 If not → program crashes ❌

---

### 🔹 Step 6: Stack Unwinding (Important)

If no handler in current function:

👉 Python goes **backwards through function calls**
👉 Checking each level for `try-except`

---

### 🔹 Step 7: Program Termination (if unhandled)

If no handler found anywhere:

👉 Python prints:

* Error type
* Message
* Traceback (line details)

---

## Exception Handling in Python

When something goes wrong at runtime, Python raises an **exception**. Without handling it, your program crashes. Exception handling lets you intercept that crash, respond gracefully, and keep going.---

## 1. The basic structure

```python
try:
    result = 10 / 0          # this raises ZeroDivisionError
except ZeroDivisionError:
    print("Can't divide by zero!")
```

Always catch **specific** exceptions — never just `except:` alone (it hides bugs).

---

## 2. Full structure — try / except / else / finally

```python
try:
    number = int(input("Enter a number: "))
    result = 100 / number

except ValueError:
    print("That's not a valid number!")   # user typed letters

except ZeroDivisionError:
    print("Can't divide by zero!")        # user typed 0

else:
    print(f"Result is {result}")          # only runs if NO exception

finally:
    print("Done.")                        # ALWAYS runs — use for cleanup
```

`else` = success path. `finally` = cleanup (close files, DB connections, etc.).

---

## 3. Catching multiple exceptions + the error message

```python
try:
    data = [1, 2, 3]
    print(data[10])           # IndexError

except (ValueError, IndexError) as e:
    print(f"Something went wrong: {e}")
    # → Something went wrong: list index out of range
```

The `as e` gives you the actual error object — very useful for logging.

---

| Exception | When it happens |
|---|---|

## 5. Custom exceptions — for your own apps

```python
class InsufficientFundsError(Exception):
    def __init__(self, balance, amount):
        self.balance = balance
        self.amount = amount
        super().__init__(f"Need ₹{amount} but only have ₹{balance}")

def withdraw(balance, amount):
    if amount > balance:
        raise InsufficientFundsError(balance, amount)
    return balance - amount

try:
    withdraw(500, 1000)
except InsufficientFundsError as e:
    print(e)   # → Need ₹1000 but only have ₹500
```

```python
    except FileNotFoundError:
        print(f"Config file '{filepath}' not found. Using defaults.")
        return {"debug": False, "port": 8000}

    except json.JSONDecodeError as e:
        print(f"Config file is corrupted: {e}")
        return None

    finally:
        print("Config load attempt complete.")

config = load_config("settings.json")
```

This is exactly how production backend code handles config loading — it never crashes the whole app just because a config file is missing.

## 7. `raise` — re-raising or throwing manually

```python
def get_age(value):
    age = int(value)           # can raise ValueError
    if age < 0:
        raise ValueError(f"Age cannot be negative: {age}")
    return age

# You can also re-raise inside an except block
try:
    get_age(-5)
except ValueError as e:
    print(f"Validation failed: {e}")
    raise    # re-raises the same exception up the call stack
```

---

## The golden rules

- Always catch specific exceptions, not bare `except:`
- Use `finally` for cleanup — closing files, releasing locks, DB connections
- Use custom exceptions to make your errors meaningful
- Never silently swallow exceptions — at minimum, log them
- `else` is underused but powerful — it's your "happy path" block
