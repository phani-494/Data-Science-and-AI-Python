# 1. What is a Tuple in Python?

A **tuple** is an **ordered, immutable collection of elements**.

**Definition:**

A tuple is a data structure that can store **multiple values in a single variable**, where the elements **cannot be changed after creation**.


# 2. Important Properties of Tuples

### 1️⃣ Ordered

Elements maintain their **insertion order**.

```
t = (5, 8, 1)
```

Index positions:

```
0 → 5
1 → 8
2 → 1
```

Access:

```python
print(t[1])
```

Output

```
8
```

---

### 2️⃣ Immutable

Once a tuple is created **its elements cannot be modified**.

Example:

```python
t = (10,20,30)

t[0] = 50
```

Error:

```
TypeError: 'tuple' object does not support item assignment
```

This means:

❌ cannot change
❌ cannot add
❌ cannot delete elements

---

### 3️⃣ Allows Duplicate Values

Tuples can contain **duplicate elements**.

```python
t = (10,20,10,30,10)
print(t)
```

Output

```
(10, 20, 10, 30, 10)
```

---

### 4️⃣ Can Store Different Data Types

Tuples can store **heterogeneous data**.

```python
t = (10, 3.14, "python", True)
print(t)
```

Output

```
(10, 3.14, 'python', True)
```

### This finds the most frequent element in the tuple.

Step-by-step understanding

t2.count(x)

Counts how many times element x appears in the tuple.

max(iterable, key=function)

max() normally returns the largest value.

But when we use key=, it decides the maximum based on the result of the function.

````python
t2=(90,87,89,90,54,65,90,34,90,-34,-56.05,89.4)
print(max(t2,key=t2.count)) 3 gives element which repeated max no of times 
````

```python
t2=(90,87,89,90,54,65,90,34,90,-34,-56.05,89.4)
print(max(t2,key=t2.count))
```

### Output

```
90
```

### Why?

This finds the **most frequent element** in the tuple.

#### Step-by-step understanding

1. **`t2.count(x)`**

   * Counts how many times element `x` appears in the tuple.

2. **`max(iterable, key=function)`**

   * `max()` normally returns the **largest value**.
   * But when we use **`key=`**, it decides the maximum based on the **result of the function**.

3. Here:

```python
key = t2.count
```

So Python does something like this internally:

| Element | t2.count(element) |
| ------- | ----------------- |
| 90      | 4                 |
| 87      | 1                 |
| 89      | 2                 |
| 54      | 1                 |
| 65      | 1                 |
| 34      | 1                 |
| -34     | 1                 |
| -56.05  | 1                 |
| 89.4    | 1                 |

The **maximum count is 4**, so the element with that count is **90**.

Therefore:

```
max(t2, key=t2.count) → 90
```

### Important Concept

This pattern is used to find the **mode (most frequent value)**.

### Clean Formula

```
max(iterable, key=iterable.count)
```

Example:

```python
l=[1,2,3,3,4,4,4,5]
print(max(l,key=l.count))
```

Output:

```
4
```

### Small Limitation

If **two elements have the same frequency**, Python returns the **first one encountered**.

Example:

```python
l=[1,2,2,3,3]
print(max(l,key=l.count))
```

Output:

```
2
```

Because **2 appears first**.

### Step-by-Step Explanation

#### 1. `sorted(t1)`

* `sorted()` **sorts elements in ascending order by default**.
* It **returns a list**, not a tuple.

So internally:

```python
sorted(t1)
```

Result:

```python
[-9, -9, -6, 0, 4, 5, 5, 6, 21, 34]
```

Notice:

* Duplicate values **remain**
* Numbers are arranged **smallest → largest**

---

#### 2. `tuple(sorted(t1))`

Since `sorted()` returns a **list**, we convert it back to a **tuple**.

```python
tuple([-9, -9, -6, 0, 4, 5, 5, 6, 21, 34])
```

Final result:

```python
(-9, -9, -6, 0, 4, 5, 5, 6, 21, 34)
```

==================================================

**=>** The Process of Defining One tuple Inside of Another tuple is called Inner OR Nested tuple.

**=>** **Syntax:** `tplobj = (Val1, Val2......(Val11, Val12....Val-1n), (Val21, Val22,...Val-2n),.....Val-n)`

**=>** Here `(Val1, Val2......,.....Val-n)` is called **Outer tuple Eleemnts**


### 1. Defining a List in a Tuple

Even though a tuple is **immutable** (cannot be changed), if one of its elements is a list, that specific list remains **mutable** (can be modified).

```python
# A tuple containing an integer, a string, and a LIST
my_tuple = (10, "Python", [1, 2, 3])

print(my_tuple[2]) # Output: [1, 2, 3]

```

### 2. The "Immutable" Rule

The tuple itself cannot be changed. You cannot add new elements to it or swap the list for a different object. However, you **can** change the contents of the list inside it.

| Action | Allowed? | Code Example |
| --- | --- | --- |
| **Change Tuple Element** | ❌ No | `my_tuple[0] = 20` (Error!) |
| **Modify List Content** | ✅ Yes | `my_tuple[2][0] = 99` (Works!) |
| **Add to List** | ✅ Yes | `my_tuple[2].append(4)` (Works!) |

---

### 3. Practical Example

Let’s look at what happens when we modify that internal list:

```python
tp = (1, 2, ["A", "B"])

# Let's modify the list at index 2
tp[2].append("C")

print(tp) 
# Output: (1, 2, ["A", "B", "C"])

```

> **Note:** Even though the tuple's content changed visually, the **identity** (memory address) of the list stayed the same, which is why Python allows it.

---

### Why use this?

You might use a List inside a Tuple when you have a fixed record (like a User Profile) where most data stays the same, but one part (like a list of recent login times) needs to be updated.

# 9. Tuple vs List Inside List

Example:

### Tuple in list

```
[(1,2),(3,4)]
```

Record is **fixed**.

---

### List in list

```
[[1,2],[3,4]]
```

Values can change.
