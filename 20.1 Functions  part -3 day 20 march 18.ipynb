# Keyword Arguments 


**=>** In some of the circumstances, we know the function name and formal parameter names and we don't know the order of formal Parameter names and to pass the data / values accurately we must use the concept of **Keyword Parameters (or) arguments**.

**=>** The implementation of Keyword Parameters (or) arguments says that all the formal parameter names used as arguments in Function call(s) as keys.

### **Syntax for function definition:-**
------------------------------------------------------------
```python
def  functionname(param1,param2...param-n):
     ---------------------------------------
     ---------------------------------------
```

### **Syntax for function call:-**
------------------------------------------------------------
```python
functionname(param-n=val-n,param1=val1,param-n-1=val-n-1,........)
```

**Here param-n=val-n, param1=val1, param-n-1=val-n-1,...... are called Keywords arguments**

**=>** When we specify Keyword arguments before Positional Arguments in Function Calls(s) then we get 

**SyntaxError: positional argument follows keyword argument because PVM  gives priority for positional arguments .**

---

### **Quick Summary of the Rule**
In Python, you can mix positional and keyword arguments, but you must follow this specific order:
1.  **Positional arguments** first.
2.  **Keyword arguments** second.



Our requirement is 

  A   B   C   D
  
 10  20   30  40
 
We want our data like this always 

- But the thing is here the developer only knows the positions(order) of parameters written in the function heading.
- The user(caller) may enter values in different positions in function calls  and he knows the name of the parameters used in the function heading but not their positions 
- so in this case , the caller must use the variables used in the function heading and can enter in any position
- Based on the keyword used , the values will be passed to their respective position here 

````python
# Function calling with Positional Arguments 
# we want always A=10 B=20,C=30,D=40
print("\tA\tB\tC\tD")
print("--"*20)
def disp(A,B,C,D):
    print("\t{}\t{}\t{}\t{}".format(A,B,C,D))
disp(10,20,30,40)
````

````python
print("\tA\tB\tC\tD")
print("--"*20)
def disp(A,B,C,D):
    print("\t{}\t{}\t{}\t{}".format(A,B,C,D))
disp(10,20,30,40)
# here 20 goes to A but we want 10 .
disp(20,30,10,40) # but here values entered in different positions and output is not proper here 
# We want order only 10  20  30  40
````

## Program for demonstrating Keyword Arguments 

````python
# so we use the keywords here and send data to their respective positions 
print("\tA\tB\tC\tD")
print("--"*20)
def disp(A,B,C,D):
    print("\t{}\t{}\t{}\t{}".format(A,B,C,D))
disp(10,20,30,40) 
# here A,B,C,D are called keyword Arguments 
# This way of calling is keyword Arguments 
disp(B=20,C=30,A=10,D=40) # now this is correct 
disp(D=40,B=20,C=30,A=10)
````

````python
print("\tA\tB\tC\tD")
print("--"*20)
def disp(A,B,C,D):
    print("\t{}\t{}\t{}\t{}".format(A,B,C,D))
disp(10,20,30,40)
disp(10,20,D=40,C=30)
# whichever going wrong, use them with their variables in function heading 
# 10,20 are the positional arguments and C,D are keyword arguments 
# First 10 goes to A ,20 goes to B and 40 goes to D as per keyword here 
````

# 🔥 What are Keyword Arguments?

👉 Keyword arguments are values passed to a function using **parameter names**.

---

## ✅ Basic Example

```python
def student(name, age):
    print(name, age)

student(name="Phani", age=22)
```

---

## 🔍 Key Idea

👉 Instead of relying on position
👉 You **explicitly tell Python** which value goes where

---

# ⚡ Why Use Keyword Arguments?

### ✅ 1. Order doesn’t matter

```python
student(age=22, name="Phani")
```

👉 Still correct ✔️

---

### ❌ Compare with positional

```python
student(22, "Phani")   # wrong mapping
```

# Variable length paramaters 

 => When we have family of multiple similar function calls with Variable number of values / arguments then with normal python programming, we must define multiple function definitions. 
 
 This process leads to more development time.

 => To overcome this process, we must use the concept of **Variable length Parameters**.

 => To Implement, Variable length Parameters concept, we must define single Function Definition and takes a formal Parameter preceded with a symbol called asterisk (**\* param**) and the formal parameter with asterisk symbol is called Variable length Parameters and whose purpose is to hold / store any number of values coming from similar function calls and whose type is **<class, 'tuple'>**.


 **Syntax for function definition with Variables Length Parameters:**

 **`def functionname(list of Positional formal params, *param1, param2=value):`**
      ____________________________________`**
    ____________________________________`**

 => Here `*param1` is called Variable Length parameter and it can hold any number of argument values (or) variable number of argument values and `*param1` type is **<class, 'tuple'>**.

 => **Rule:-** The `*param1` must always written at last part of Function Heading and it must be only one (but not multiple).

 => **Rule:-** When we use Variable length and default parameters in function Heading, we use default parameter as last and before we use Variable length parameter and in function calls, we should not use default parameter as Key word argument bcoz Variable number of values are treated as Positional Argument Value(s).

---

### **Key Takeaways & Clarification**

* **The Asterisk (`*`):** This is the "magic" part. It tells Python to pack all remaining positional arguments into a single tuple.
* **Data Type:** Even if you pass only one value, or zero values, the variable inside the function will always be a **tuple**.
* **Positioning :** In standard practice, `*args` (the variable-length param) usually comes after fixed positional arguments but before keyword-only arguments or those with default values to avoid ambiguity.

# 🔥 What are Variable-Length Arguments?

👉 They allow a function to accept **any number of inputs**
👉 You don’t need to know how many arguments will come

---

# 🧱 Two Types

### 1️⃣ `*args` → multiple **positional arguments**

### 2️⃣ `**kwargs` → multiple **keyword arguments**

---

# 🚀 1️⃣ `*args` (Non-keyword variable arguments)

## 📌 Concept

* `*args` collects values into a **tuple**

---

## ✅ Basic Example

```python
def add(*nums):
    print(nums)

add(1, 2, 3, 4)
```

### 🔍 Output

```python
(1, 2, 3, 4)
```

---

## 🔥 Real Use (Important)

```python
def total(*nums):
    s = 0
    for i in nums:
        s += i
    return s

print(total(10, 20))
print(total(5, 10, 15, 20))
```

👉 Works for any number of inputs

---

## ⚡ Execution Flow

```python
total(5, 10, 15)
```

👉 Internally:

```python
nums = (5, 10, 15)
```

---

## 💼 Real-Time Example

### 🧾 Billing system

```python
def bill(*items):
    print("Total:", sum(items))

bill(100, 200, 50)
bill(500, 300)
```

👉 Number of items is unknown → perfect use case

---

# 🚀 2️⃣ `**kwargs` (Keyword variable arguments)

## 📌 Concept

* Collects values into a **dictionary**

---

## ✅ Basic Example

```python
def details(**data):
    print(data)

details(name="Phani", age=22)
```

### 🔍 Output

```python
{'name': 'Phani', 'age': 22}
```

---

## 💼 Real-Time Example

### 👤 User profile system

```python
def profile(**info):
    for k, v in info.items():
        print(k, ":", v)

profile(name="Phani", city="Bangalore", age=22)
```

# 🔥 Combining Everything

```python
def demo(a, *args, **kwargs):
    print("a =", a)
    print("args =", args)
    print("kwargs =", kwargs)

demo(10, 20, 30, x=100, y=200)
```

---

## 🔍 Output

```python
a = 10
args = (20, 30)
kwargs = {'x': 100, 'y': 200}
```

# ⚠️ Important Rules (VERY IMPORTANT)

---

## 📌 Order of parameters

```python
def func(a, b=10, *args, **kwargs):
    pass
```

👉 Always follow:

```text
1. Positional
2. Default
3. *args
4. **kwargs
```

---

## ❌ Wrong Order

```python
def func(*args, a):
    pass
```

👉 ❌ Error
