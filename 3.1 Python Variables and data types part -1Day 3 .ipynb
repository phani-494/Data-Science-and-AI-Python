##  What is Data in Python?

In Python, **data** means:

> Any value or information that a program can store, process, and manipulate.

Simply put, **data is the content your program works with.**

---

###🔹 Example of Data

```

Python stores data in different **data types**:

### 1️⃣ Numeric Data

* `int` → 10, 25

### 2️⃣ Text Data

##  How Inputs Are Stored in Python?

In Python, when a user gives input, Python:

1️⃣ Takes the input as a **string**
2️⃣ Creates an **object in memory (heap)**
3️⃣ Stores a **reference** to that object in a variable

---

## 🔹 Step-by-Step Example

name = input("Enter your name: ")
```

### 🔸 What Happens Internally?

1. User types: `Phanindra`
2. `input()` reads it
3. Python creates a **string object** `"Phanindra"` in heap memory
4. Variable `name` stores the reference to that object

📌 By default, `input()` always returns **string type**.

---

## 🔹 Example with Numbers

age = input("Enter age: ")
```

If user types: `25`

Python stores:

```
"25"  → string (not integer)

## 🔹 Memory Concept

When you write:

x = input()
```

Internally:

```
User Input
    ↓
String Object Created (Heap)
    ↓
Variable x stores reference
```

📌 In Python:

* Objects → stored in Heap
* Variable → reference to object
* Function calls → stored in Stack

---

## 🔹 Important Concept 

👉 Python does **not store raw input directly in variable memory location** like C.
👉 It creates an object and variable points to that object.

Everything in Python is an **object**.

---

## 🧠 Short  Definition

> In Python, user input is taken as a string by default, stored as an object in heap memory, and the variable stores a reference to that object.

## 🐍 Identifiers in Python

In Python, an **identifier** is:

> The name given to a variable, function, class, module, or any other object.

It is used to **identify** a particular entity in a program.

---

## 🔹 Simple Example

x = 10
name = "Phanindra"

def add(a, b):
    return a + b
```

Here:

* `x` → identifier
* `name` → identifier
* `add` → identifier
* `a`, `b` → identifiers

---

## 🔹 Rules for Naming Identifiers

1️⃣ Must start with:

* A letter (a–z, A–Z)
* Or underscore `_`

2️⃣ Cannot start with a number
❌ `1name` → Invalid

3️⃣ Can contain:

* Letters
* Digits
* Underscore

4️⃣ Cannot use special symbols
❌ `name@123` → Invalid

5️⃣ Cannot use Python keywords
❌ `if`, `for`, `while` → Invalid

---

## 🔹 Examples

### ✅ Valid Identifiers

total_marks
_count
name1
```

### ❌ Invalid Identifiers

1value      # starts with number
total-marks # contains hyphen
class       # keyword
```

---

## 🔹 Case Sensitivity

Python is case-sensitive:

age = 20
Age = 30
```

`age` and `Age` are different identifiers.

## Importance of Identifiers OR Variables
========================================================================

=> In Any Programming Language, Literals Must be Stored in Main Memory by allocating Sufficient Amount of Memory space with the Help Of Data Types.

=> We know that All types Literals are stored in main memory by having memory space.

=> To Process values which are present in memory space, Programmers must give DISTINCT NAMES to the created memory spaces. These DISTINCT NAMES makes us to identify the values present in memory space and hence they are called IDENTIFIERS.

=> The IDENTIFIER Values are Changing/ Varying during Program Execution and hence IDENTIFIERS are also called VARIABLES.

=> Hence All types of LITERALS are stored in the form VARIABLES and all Variables are called OBJECTS.

### Definition of Variable:

A Variable is an Identifier, whose Value can be Changed OR Varying During the Program Execution.

### Rules for Using Variables OR Identifiers in Python
To use variables in a Python program, you must follow specific naming rules:

RULE-1: The variable name is a combination of alphabets, digits, and the special symbol underscore (_).

RULE-2: The first letter of a variable name must start with either an alphabet OR the underscore (_) symbol.

Examples:

123sal=23 --------- Invalid

_sal=45 ----------- Valid

sal12=34 ---------- valid

__sal=45 ---------- valid

_1=12 ------------- valid

RULE-3: Within the variable name, special symbols are not allowed except under score ( _ ).

Examples:

emp sal=34 -------- Invalid

emp_sal=45 -------- valid

emp-sal=45 -------- invalid

emp$sal=45 -------- Invalid

## RULE-4:
No Keywords are to be used as Variable Names because Keywords are reserved words that provide special meaning to the language compilers.

while=45 — Invalid

while1=45 — Valid

if=34 — Invalid

_if=45 — Valid

True=56 — Invalid

true=56 — Valid

## RULE-5:
All variable names in Python are Case Sensitive.

Example: If you define age=99, AGE=98, Age=97, and aGe=96, Python treats them as four distinct variables.

print(age, AGE, Age, aGe) would output: 99 98 97 96
