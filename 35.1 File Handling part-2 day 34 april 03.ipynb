# 📂 What are File Modes in Python?

👉 **File modes specify how you want to open a file (read, write, append, etc.)**

---

## 🔹 1. `"r"` → Read Mode

## ✅ Purpose

👉 Read data from file

---

## ⚙️ Behavior

* File **must exist**
* Starts reading from beginning

---

## 💻 Example

```python
with open("data.txt", "r") as f:
    print(f.read())
```

---

## ❌ Error

👉 If file not present → `FileNotFoundError`

---

## ✅ Purpose

👉 Write data to file

---

## ⚙️ Behavior

---

## 💻 Example

```python
with open("data.txt", "w") as f:
    f.write("Hello")
```

---

## ⚠️ Important

👉 Old data is **deleted**

---

---

## 🔹 3. `"a"` → Append Mode

## ✅ Purpose

👉 Add data to existing file

---

## ⚙️ Behavior

* Creates file if not exists
* Adds data at **end**

---

## 💻 Example

```python
with open("data.txt", "a") as f:
    f.write("New line\n")
```

---

---

## 🔹 4. `"r+"` → Read + Write

## ✅ Purpose

👉 Read and write both

---

## ⚙️ Behavior

* Cursor starts at beginning

---

---

## 🔹 5. `"w+"` → Write + Read

## ✅ Purpose

👉 Write and read

---

## ⚙️ Behavior

* Overwrites file
* Creates if not exists

---

---

## 🔹 6. `"a+"` → Append + Read

## ✅ Purpose

👉 Append and read

---

## ⚙️ Behavior

* File created if not exists
* Cursor at end

# 📂 `"r"` Mode in Python (Read Mode)

## ✅ Simple Definition

👉 `"r"` mode is used to **open a file for reading data**

---

# 🧠 Core Behavior

When you do:

```python
with open("data.txt", "r") as f:
    data = f.read()
```

👉 Python:

1. Opens the file
2. Places cursor at **beginning**
3. Reads content into memory

---

# ⚙️ Important Rules

## 🔹 1. File MUST exist

👉 If file is not present:

❌ Error:

```text
FileNotFoundError
```

---

## 🔹 2. Read Only

👉 You **cannot write** in `"r"` mode

❌ This will fail:

```python
f.write("Hello")  # Error
```

---

## 🔹 3. Cursor starts at beginning

```text
|Hello World|
 ^
 cursor here
```

---

# 💻 Common Reading Methods

---

## 🔹 1. `read()`

👉 Reads entire file

```python
with open("data.txt", "r") as f:
    print(f.read())
```

---

## 🔹 2. `readline()`

👉 Reads one line

```python
print(f.readline())
```

---

## 🔹 3. `readlines()`

👉 Reads all lines as list

```python
print(f.readlines())
```

---

# 🌊 Internal Working (Important)

👉 When using `"r"` mode:

```text
Disk → Bytes → Buffer → Python String Object
```

✔ Data is loaded into memory
✔ Then you can process it

# 💼 Real Data Engineering Use

👉 `"r"` mode is used for:

* Reading CSV files

👉 Always check file exists

# 🧠 Final Master Line

👉 **"r" mode = Read existing file from start**

---




### **1. r (Read Mode)**
* **=>** This Mode is used for Opening  the File in  **Read Mode** and we can perform **Read Operation**.
* **=>** If we don't Specify any File Opening Mode then **PVM** (Python Virtual Machine) by default Takes as **"r" mode** (default file mode).
* **=>** If we open the file name in **"r" mode** and if that file name **does not exist** then we get **FileNotFoundError**.

---

### **2. w (Write Mode)**
* **=>** This Mode is used for **Creating the File** and Open that File in **Write Mode** and we can perform **Write Operation**.
* **=>** When we Choose **NEW FILE** and If we open in **"w" mode** then **NEW FILE Created** and Opened in write Mode and we can perform Write Operations.
* **=>** When we Choose **EXISTING FILE** and If we open in **"w" mode** then **EXISTING FILE** Opened in write Mode and Existing Data **OVERLAPPED** (overwritten/truncated) with New Data.



### **Summary Key Points:**
* **Default:** Python assumes you want to read (`r`) if you don't tell it otherwise.
* **Safety:** `r` mode is safe (it won't delete anything), but it crashes if the file is missing.
* **Danger Zone:** `w` mode is "destructive" for existing files because it wipes the old content (overlaps) as soon as you open it.

4. Saves it to disk

```text
Old Data → ❌ Deleted
New Data → Written
```

```text
Python
```

## 🔹 4. Cursor Behavior

### **3. a (Append Mode)**
* **=>** This Mode is used for **Creating the File** and Open that File in **Write Mode** and we can perform **Write Operation**.

* **=>** This Mode is used for **Opening the File in Read Mode** and First we can perform **Read Operation** and Later we can Perform **Write Operation** also.

### **Key Takeaways :**

* **Append vs. Write:** While both `w` and `a` can create a new file, the difference lies in how they handle existing files. `w` deletes old data (Overlaps), while `a` keeps old data and adds new data to the end.
* **The Power of `r+`:** This is the first mode mentioned that allows both reading and writing. However, it is strict—the file **must** already exist on your computer, or the program will crash.

When using these "plus" modes (`w+`, `a+`, `r+`), remember that after you write something, the "file pointer" is at the end of what you just wrote. If you want to read the file immediately, you usually need to use **`file.seek(0)`** to move the pointer back to the beginning!

* **=>** This mode is used for **Opening the File in Write Mode ONLY ONCE** and First we can perform **Write Operation** and Later we can **Perform Read Operation also**.
* **=>** If we open an existing file in **"x+" mode** then we get **FileExistsError**.

### **Summary of the "x" Modes**
The `x` mode is the "safety first" mode.

* Unlike `w` (which deletes your old data) or `a` (which adds to it), **`x` refuses to work if the file already exists.**
*
* * It is perfect for when you want to ensure you are creating a **brand new file** and don't accidentally overwrite something important.

In Python, **File Opening Modes** tell the interpreter how you intend to interact with a file—whether you want to just read it, write new data, or add to what’s already there.


---

### 1. Basic Modes (Single Purpose)
* **`r` (Read):** The default mode. It opens a file for reading. If the file does not exist, it raises a `FileNotFoundError`.
* **`w` (Write):** Opens a file for writing. **Warning:** If the file already exists, it truncates (erases) everything inside first. If it doesn't exist, it creates a new one.
* **`a` (Append):** Opens a file for writing, but starts at the very end of the file. It does **not** erase existing data.
* **`x` (Exclusive Creation):** Used specifically to create a new file. If the file already exists, this operation fails (raises a `FileExistsError`).

### 2. Plus Modes (Dual Purpose / Updating)
The `+` symbol generally means "updating," allowing you to both read and write with the same file handle.

* **`r+` (Read + Write):** Opens the file for both reading and writing. The file pointer starts at the beginning. It does not delete the file content automatically, but writing will overwrite existing characters.
* **`w+` (Write + Read):** Like `w`, it erases the file first. However, it also allows you to read back what you just wrote.
* **`a+` (Append + Read):** Like `a`, it opens for appending, but also lets you read the file. The "write" pointer is always forced to the end of the file.
* **`x+` (Create + Read/Write):** Like `x`, it creates a new file and allows you to read/write to it. It fails if the file already exists.

### Quick Comparison Table

| Mode | Read | Write | Create File? | Overwrites Existing? | Pointer Position |
| :--- | :--- | :--- | :--- | :--- | :--- |
| **`r`** | Yes | No | No | No | Beginning |
| **`w`** | No | Yes | Yes | **Yes (Truncates)** | Beginning |
| **`a`** | No | Yes | Yes | No | End |
| **`r+`** | Yes | Yes | No | No (But overwrites chars) | Beginning |
| **`w+`** | Yes | Yes | Yes | **Yes (Truncates)** | Beginning |
| **`a+`** | Yes | Yes | Yes | No | End |
