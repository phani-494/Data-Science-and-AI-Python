## Compiler vs Interpreter 

### 5. Interpreted Programming
 * When we develop any Python program, we must give some file name with an extension .py (FileName.py).
 * When we execute a Python program, two processes take place internally:
   * Compilation Process
   * Execution Process
The Compilation Process
In the COMPILATION PROCESS, the Python Source Code is submitted to the Python Compiler. It reads the source code and checks for errors by verifying syntax.
 * If no errors are found, the Python Compiler converts the code into an Intermediate Code called BYTE CODE with an extension .pyc (FileName.pyc).
 * If errors are found in the source code, then the error is displayed on the console.
The Execution Process
In the EXECUTION PROCESS, the PVM reads the Python Intermediate Code (Byte Code) line by line and converts it into Machine Understandable Code (Executable or Binary Code). This is then read by the OS and Processor to finally give the result.
> Conclusion: In Python program execution, the Compilation Process and Execution Process take place through line-by-line conversion. This makes it an Interpretation Based Programming Language.
> 
Definition of PVM (Python Virtual Machine)
PVM is a program within the Python software whose role is to read LINE by LINE of Byte Code and convert it into Machine Understandable Code (Executable or Binary Code)

## 🔹 3️⃣ Dynamic Typing with Safety

```python
x = "10"
y = 5
print(x + y)  # Error detected safely
```

#  Types of Errors in Python

In Python, errors are mainly classified into **three types**:

---

## 🔹 1️⃣ Syntax Errors (Compile-Time Errors)

These occur when you **violate Python grammar rules**.

### ❌ Example:

```python
if x > 10
    print("Hello")
```

### 🔎 Problem:

Missing `:` after condition.

### ✅ Fix:

```python
if x > 10:
    print("Hello")
```

📌 Detected before execution starts.

---

## 🔹 2️⃣ Runtime Errors (Exceptions)

These occur **during program execution**.

### ❌ Example:

```python
x = 10 / 0
```

### 🔎 Error:

`ZeroDivisionError`

Common runtime errors:

* `ZeroDivisionError`
* `TypeError`
* `ValueError`
* `IndexError`
* `KeyError`
* `FileNotFoundError`

### ✅ Handled using:

```python
try:
    x = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero")
```

---

# 🔹 3️⃣ Logical Errors

These occur when:

* Program runs
* No error message
* But output is wrong

### ❌ Example:

```python
a = 10
b = 20
print("Sum:", a - b)   # Wrong logic
```

Output: `-10`
Expected: `30`

📌 Hardest to detect
📌 Found by testing & debugging

---

# 🐍 Why Python is Extensible?

Python is called **extensible** because we can **extend its functionality by adding code written in other languages like C or C++**.

👉 In simple words:

> Extensible means we can add new features to Python using other programming languages.

---

## 🔹 1️⃣ Adding C/C++ Code to Python

Python allows integration with:

* C
* C++

If some operation is:

* Very slow in Python
* Performance-critical

We can write that part in C/C++ and connect it to Python.

---

## 🔹 2️⃣ Why This Is Useful?

Python is:

* Easy to write
* But sometimes slower than C

So developers:

* Write main program in Python
* Write heavy computations in C
* Combine both

This improves performance.

---

## 🔹 3️⃣ Real-World Example

Libraries like:

* NumPy
* TensorFlow

Are written internally in C/C++ but used in Python easily.

That’s extensibility in action 🔥

---

## 🔹 4️⃣ How Python Supports Extensibility?

Because:

* Python has a C API
* CPython itself is written in C
* External modules can be linked

(Reference implementation: CPython)

---

## 🔥 Simple Example Concept

```python
# Python calls a C function internally
import math
print(math.sqrt(16))
```

The `math` module uses C code internally for speed.

---

## 🧠 Short Definition

> Python is extensible because it allows integration of code written in other languages like C and C++ to enhance functionality and performance.

---

## ⚡ Difference Between Extensible and Extendable

* **Extensible** → Add external code (like C modules)
* **Extendable** → Add new Python modules/classes
