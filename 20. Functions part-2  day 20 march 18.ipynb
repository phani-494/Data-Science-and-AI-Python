## take a list of values and find sum and average 

````python
def ip_val1():
    l1=[]
    while(True):
        ip=input("Enter the values and press @ to stop :")
        if ip=="@":
            break
        l1.append(float(ip))
    return l1
def cal1():
    l1=ip_val1()
    sum1=0
    for i in l1:
        sum1=sum1+i
    a=sum1/len(l1)
    return sum1,a
s,a=cal1()
print("sum={},average={}".format(s,a))
````

````python
# Function to calculate sum
def find_sum(lst):
    total = 0
    for i in lst:
        total += i
    return total

# Function to calculate average
def find_avg(lst):
    return find_sum(lst) / len(lst)


numbers = []

while True:
    val = input("Enter number (@ to stop): ")
    
    if val == '@':
        break
    
    # Check if input is a valid number
    try:
        num = float(val)
        numbers.append(num)
    except ValueError:
        print("Invalid input! Please enter a number.")


# Check if list is empty
if len(numbers) == 0:
    print("No valid numbers entered")
else:
    print("Sum =", find_sum(numbers))
    print("Average =", find_avg(numbers))
````

## Parameters and Arguments 

 ## Parameters in Python

In Python, parameters are classified into two types:

1.  **Formal Parameters / Variables**
2.  **Local Variables / Parameters**

---

### 1. Formal Parameters / Variables

* **Definition:** Formal Parameters / Variables are those which are used in the **Function Heading**.
* **Purpose:** The purpose of Formal Parameters is to **"store OR Hold the Inputs coming From Function Calls"**.
* **Access:** The values of Formal Parameters can be accessed within the corresponding Function Definition, but it is **not possible** to access them in other parts of the Function Definition OR in other parts of the Program.

---

### 2. Local Variables / Parameters

* **Definition:** Local Variables / Parameters are those which are used as a **part of the Function Body**.
* **Purpose:** The purpose of Local Variables is to **"store the Result of Function Processing Logic"**.
* **Access:** The values of Local Parameters can be accessed within the corresponding Function Definition, but it is **not possible** to access them in other parts of the Function Definition OR in other parts of the Program.

---

**def sumop(a,b,c): # Here a,b,c are called Formal Parameters**

      d=a+b+c # Here d is called local Variable**
      
**sumop(10,20,30) # Function call--here 10 20 30 are called Argument Values**

## Arguments in Python

* **Definition:** Arguments are also called **Variables**.
* **Usage:** Arguments are Variables used **inside of a Function Call**.

---

### Examples:

1.  **Direct Values:**
    `sumop(10, 20, 30)`
    *# Function call—here 10, 20, and 30 are called **Argument Values**.*

2.  **Using Variables:**
    `x, y, z = 10, 20, 30`
    `sumop(x, y, z)`
    *# Function call—here x, y, and z are called **Arguments**.*

---

### Relationship Between Arguments and Parameters:
The relationship between Arguments and Parameters is that **Every Value of an Argument Must pass to Formal Parameters**.

---


## Formal Parameters vs. Actual Arguments

| Feature | Formal Parameters | Actual Arguments (or Argument Values) |
| :--- | :--- | :--- |
| **Definition** | Variables listed in the function header. | The real values passed to the function when it is called. |
| **Purpose** | They act as **placeholders** for the data the function will receive. | They are the **actual data** the function will process. |
| **Scope** | They only exist inside the function (Local Scope). | They exist in the part of the program where the function is called. |
| **Example** | In `def sumop(a, b, c):`, **a, b,** and **c** are formal parameters. | In `sumop(10, 20, 30)`, **10, 20,** and **30** are actual arguments. |

---

### The Workflow of a Function Call
When you run `sumop(10, 20, 30)`, a process called **parameter passing** occurs:
1.  The value **10** is assigned to parameter **a**.
2.  The value **20** is assigned to parameter **b**.
3.  The value **30** is assigned to parameter **c**.

# Default parameters 

### **2) Default Parameters (or) arguments**

-  When there is a Common Value for family of Similar Function Calls then Such type of Common Value(s) must be taken as
  default parameter with common value (But not recommended to pass by using Posstional arguments OR Parameters)
-----------------------------------------------------------------------------------------
**Syntax for Function Definition with Default Parameters**

```python
def functionname(param1, param2, .... param-n-1=Val1, Param-n=Val2):
    ---------------------------------------------------------
    ---------------------------------------------------------
```
Here **param-n-1** and **param-n** are called **"default Parameters"**.
and **param1, param-2...** are called **"Possitional parameters"**.

- Rule-:When we use default parameters in the function definition, They must be used as
last Parameter(s) otherwise we get Error( SyntaxError: non-default argument (Possitional ) follows default argument).

---

##### **Quick Summary of the Rule**
In Python, you always have to put your "required" (positional) arguments first. 
Think of it like a guest list: you name the specific VIPs first, and the "plus-ones" (defaults) come after.

**Correct:** `def greet(name, message="Hello"):`

**Incorrect:** `def greet(message="Hello", name):` (This will trigger the **SyntaxError** mentioned in your notes).



# 🔥 What are Default Parameters?

👉 Default parameters are values **assigned in the function definition**
👉 If the user does NOT pass a value → default is used

---

## ✅ Basic Example

```python
def greet(name="Guest"):
    print("Hello", name)

greet()          # uses default
greet("Phani")   # overrides default
```

---

## 🔍 Output

```
Hello Guest
Hello Phani
```

---

# ⚡ Execution Flow

1. Function defined → `name="Guest"` stored
2. If no argument → use `"Guest"`
3. If argument given → replace default

# ❗ Important Rule (VERY IMPORTANT)

## ✅ Default parameters must be at the end

```python
def test(a, b=10):
    pass
```

## ❌ Wrong

```python
def test(a=10, b):
    pass
```

👉 ❌ SyntaxError
👉 Python doesn’t allow this

# ⚠️ Common Mistakes

---

## ❌ Mutable Default Argument (VERY IMPORTANT 🔥)

```python
def add_item(item, lst=[]):
    lst.append(item)
    return lst

print(add_item(1))
print(add_item(2))
```

👉 Output:

```
[1]
[1, 2]   ❌ unexpected!
```

👉 Why?
Default list is **shared across calls**

---

## ✅ Correct Way

```python
def add_item(item, lst=None):
    if lst is None:
        lst = []
    lst.append(item)
    return lst
```
