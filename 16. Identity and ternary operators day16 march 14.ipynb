## Identity Operators in Python

**Purpose:** The purpose of Identity Operators is **"To compare the Memory Address of Two Objects."** 
                                                    
In Python Programming, we have 2 Identity Operators. They are:

1. **is** Operator
2. **is not** Operator

---

### 1. is Operator

**Syntax:** `Object1 is Object2`

* **Returns True:** Provided the Memory Address of Object1 and Object2 are **Same**.
* **Returns False:** Provided the Memory Address of Object1 and Object2 are **Different**.

---

### 2. is not Operator

**Syntax:** `Object1 is not Object2`

* **Returns True:** Provided the Memory Address of Object1 and Object2 are **Different**.
* **Returns False:** Provided the Memory Address of Object1 and Object2 are **Same**.

---


### Examples:

1. **NOTE 1:** All Deep Copy Objects with **"is"** Operator returns **True**
2. **NOTE 2:** All Deep Copy Objects with **"is not"** Operator returns **False**
3. **NOTE 3:** All Shallow Copy Objects with **"is"** Operator returns **False**
4. **NOTE 4:** All Shallow Copy Objects with **"is not"** Operator returns **True**

---

````python
l1=[1,2,3]
l2=l1 # deep copy -so both will have the same memory address here 
print(id(l1),id(l2))
print(l1 is l2)
print(l1 is not l2)
````

````python
l1=[1,2,3]
l2=l1.copy() # shallow copy -so both will not have the same memory address here 
print(id(l1),id(l2))
print(l1 is l2)
print(l1 is not l2)
````

````python
a=None
b=None
print(a is b)
print(a is not b)
````

````python
a=1
b=1 # same value is present in both objects here . so same id for both a and b
print(a is not b)
print(a is b)
````

## Identity Operators in Python

**Identity operators** are used to check **whether two variables refer to the same object in memory**.

Python has **two identity operators**:

| Operator | Meaning                                               |
| -------- | ----------------------------------------------------- |
| `is`     | True if both variables refer to the **same object**   |
| `is not` | True if both variables refer to **different objects** |

---

# 1️⃣ `is` operator

Checks whether **two variables point to the same memory object**.

### Example

```python
a = [1,2,3]
b = a

print(a is b)
```

Output

```
True
```

Explanation:

```
      [1,2,3]
      /     \
     a       b
```

Both variables refer to **one object**.

---

# 2️⃣ Different objects with same value

```python
a = [1,2,3]
b = [1,2,3]

print(a is b)
```

Output

```
False
```

Because Python created **two separate list objects**.

```
a → [1,2,3]  (Memory A)
b → [1,2,3]  (Memory B)
```

---

# 3️⃣ `is not` operator

Checks if objects are **not the same**.

```python
a = [1,2,3]
b = [1,2,3]

print(a is not b)
```

Output

```
True
```

---

# 4️⃣ Difference between `is` and `==`

| Operator | Checks          |

| `==`     | Value equality  |
| `is`     | Memory identity |

Example:


# 1️⃣ Because each dictionary literal creates a new object

Every time Python executes a **dictionary literal `{}`**, it **creates a new dictionary object in memory**.

So Python runs the code like this step-by-step:

### Step 1

```python
d1 = {1:123,2:124}
```

Python creates a **dictionary object in memory**

```
Memory A
{1:123,2:124}
```

and `d1` points to it.

```
d1 ───► Memory A
```

---

### Step 2

```python
d2 = {1:123,2:124}
```

Python again sees `{}` and **creates another new dictionary object**.

```
Memory B
{1:123,2:124}
```

Now:

```
d1 ───► Memory A
d2 ───► Memory B
```

Even though the **values are identical**, the **objects are different**.

---

# 2️⃣ That is why `is` returns False

`is` checks **memory identity**.

```
d1 is d2
```

means:

```
Memory A == Memory B ?
```

Answer:

```
False
```


Output

```
True
```

Because `==` compares **data inside objects**, not memory location.

# 5️⃣ Important rule in Python

Python **does not automatically merge mutable objects** like:

* dictionaries
* lists
* sets

Because they can **change later**, and sharing them would cause bugs.

Example:

```python
d1 = {1:100}
d2 = {1:100}

d1[2] = 200

print(d2)
```

If Python reused the same object, `d2` would also change — which is **dangerous**.

---

✅ **Key Python concept**

| Object Type                                   | Memory Reuse      |
| --------------------------------------------- | ----------------- |
| Immutable (small integers, strings sometimes) | Sometimes reused  |
| Mutable (list, dict, set)                     | Always new object |

In Python, when you create two separate tuples like `t1 = (1, 2, 3)` and `t2 = (1, 2, 3)`, 
the interpreter generally treats them as **distinct literal declarations**.

Even though the content is identical, Python allocates a new chunk of memory for each one.
    Since the `is` operator checks for **Identity** (do they point to the exact same spot in RAM?), it returns `False`.

### Why are two different objects created?

1. **Memory Independence:** By default, every time you define a collection (like a list or a tuple), 
Python creates a new object to ensure that operations on one don't accidentally affect another (though this is more
critical for mutable objects like lists).

2. **The Cost of "Interning":** For Python to make `t1 is t2` return `True`, it would have to search through 
all existing objects in memory every time you create a new one to see if a matching tuple already exists. 
    This would make the language significantly slower.

3. **Immutability vs. Identity:** While tuples are immutable (you can't change them), Python only "interns" (reuses) very specific, simple objects like small integers (usually -5 to 256) and certain short strings to save memory. Larger structures like your tuple are typically not automatically reused.


While Python generally creates new objects for collections (like your previous tuple example), 
it handles certain strings differently to save memory and speed up comparisons.

### Why do these strings share an address?

1. **The Interning Mechanism:** Python maintains an internal table (a "pool") of certain strings. 
    When you create a new string literal that looks like a "name" (identifiers consisting of letters, numbers, or underscores), 
Python checks if it already exists in that pool. If it does, it simply points the new variable to the **existing** memory address.
    
2. **Efficiency:** Comparing strings by value (`==`) requires checking them character by character. Comparing them by identity (`is`) is nearly instantaneous because it only compares two memory addresses (integers). By interning common strings, Python makes your code run faster.

    3. **Predictable Literals:** Since strings are **immutable** (you can't change "Indra" to "Indri" without creating a whole new string), it is perfectly safe for `s1` and `s2` to point to the same spot. Neither variable can "corrupt" the data for the other.

---

### When does Interning *not* happen?

Python is picky about what it interns automatically. It usually only happens for:

* **Small Integers:** Between **-5** and **256**.
* **Simple Strings:** Strings that look like valid Python identifiers (no spaces or special symbols).

**Try this test to see the difference:**

```python
# Simple strings (Interned)

b = "hello"
print(a is b)  # True

# Strings with spaces/symbols (Usually NOT interned)
c = "hello world!"
d = "hello world!"
print(c is d)  # False (in most interactive shells)

```

* Complex numbers

Because caching these objects **would not give much performance benefit**.

## Why both have the same address?

Here Python performs an optimization called **constant pooling / constant folding** during compilation.

### What happens internally

When Python compiles this line:

```python
a, b = 1.2, 1.2
```

it sees that **both constants are the same** and stores **one float object** in the program’s constant table.

Conceptually:

```
      1.2  (one object in memory)
      / \
     a   b
```

So both variables point to the **same object**, therefore:

```
a is b → True
```

---

## But this is not guaranteed

If the float values are created **separately at runtime**, Python will create **two different objects**.

Example:

```python

b = float("1.2")

print(a is b)
```

Output:

```
False
```

Memory view:

```
a → 1.2  (memory A)
b → 1.2  (memory B)
```

## Ternary Operator in Python

The **ternary operator** is used to write **conditional expressions in one line**.

It is a **short form of `if-else`**.

### Syntax


## Python Ternary Operator

**The Name of Python Ternary Operator is "if..else" Operator.**

### Syntax:

`varname = Expression-1 if Test Condition else Expression-2`

---

### Key Points:

* **Test Condition:** This can be either a **Relational** OR **Logical Expression**, and its result will be either **True** or **False**.
* **True Result:** If the Test Condition result is **True**, then the **PVM** (Python Virtual Machine) executes **Expression-1**, and the result is placed in the LHS (Left-Hand Side) **Varname**.
* **False Result:** If the Test Condition result is **False**, then the **PVM** goes to the **else** part and executes **Expression-2**, and the result is placed in the LHS **Varname**.
* **Conclusion:** In the Python Ternary Operator, at any point in time, the PVM executes either **Expression-1** OR **Expression-2**, and the resulting value is assigned to the LHS variable.

---
