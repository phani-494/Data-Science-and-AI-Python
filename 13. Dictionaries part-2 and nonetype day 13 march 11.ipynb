# 2️⃣ Understanding `max()`

Normally:

```python
max(products)
```

This returns the **largest key alphabetically**, not the value.

Example:

Because `"os"` is largest alphabetically.

# 3️⃣ Important Concept: `key=products.get`

```python
max(products, key=products.get)
```

Here Python does this internally:

```
products.get("os")    → 56000
products.get("apple") → 45609
products.get("berry") → 90432
```

Then `max()` compares **values**, not keys.

| Key   | Value |
| ----- | ----- |
| os    | 56000 |
| apple | 45609 |
| berry | 90432 |

Maximum value = **90432**

Corresponding key = **"berry"**

---

# 4️⃣ Final Output

```
berry
```

# 5️⃣ How Python Internally Works

Python checks like this:

```
max(["os","apple","berry"], key=products.get)
```

Step-by-step:

```
products.get("os")    → 56000
products.get("apple") → 45609
products.get("berry") → 90432
```

Highest value → **90432**

So result → **"berry"**

## Dictionary inside Set in Python

❗ **Important Rule**

A **dictionary cannot be stored inside a set**.

Because:

| Data Type  | Mutable / Immutable |
| ---------- | ------------------- |
| Dictionary | Mutable             |
| Set        | Mutable             |

And **sets only allow immutable (hashable) elements**.

So Python will give an error.

---

# 1️⃣ Example (Error Case)

```python
s = { {"name": "Ravi"}, {"name": "Anu"} }

print(s)
```

Output

```
TypeError: unhashable type: 'dict'
```

Reason:

```
Set elements must be hashable
Dictionary is not hashable
```

---

# 2️⃣ Why Dictionary is Not Allowed in Set

Set internally uses **hashing**.

Only **hashable objects** can be stored in a set.

Examples of **hashable types**:

* int
* float
* string
* tuple (if immutable)

Example:

```python
s = {10, "python", (1,2,3)}

print(s)
```

This works.

---

# 3️⃣ Why Dictionary is Unhashable

Dictionary is **mutable**.

Example:

```python
d = {"a":10}
d["a"] = 20
```

Since values can change, Python **cannot compute a fixed hash value**.

So it **cannot be stored in a set**.

---

# 4️⃣ How to Store Dictionary-like Data in Set

### Method 1: Convert dictionary to tuple

```python
s = {tuple({"name":"Ravi"}.items())}

print(s)
```

Output

```
{(('name', 'Ravi'),)}
```

---

### Method 2: Use frozenset

```python
s = {frozenset({"name":"Ravi"}.items())}

print(s)
```

Output

```
{frozenset({('name', 'Ravi')})}
```

Because **frozenset is immutable and hashable**.

---

# 5️⃣ Important Concept

| Structure  | Allowed in Set |
| ---------- | -------------- |
| List       | ❌              |
| Dictionary | ❌              |
| Tuple      | ✔              |
| String     | ✔              |
| Frozenset  | ✔              |

---

# 6️⃣ Interview Question Concept

Why **dict cannot be element of set**?

Answer:

```
Because dictionary is mutable and unhashable,
while set elements must be immutable and hashable.
```

---

✅ **Final Rule to Remember**

```
Mutable objects cannot be stored inside a set.
```

Examples of **mutable objects**:

* list
* dictionary
* set

## `None` in Python

`None` is a **special value in Python** that represents **"no value" or "nothing"**.

It means the variable **does not contain any meaningful data**.

---

# 1️⃣ What is `None`?

```python
x = None
print(x)
```

Output

```
None
```

So:

```text
None → absence of value
```

---

# 2️⃣ Type of `None`

```python
print(type(None))
```

Output

```
<class 'NoneType'>
```

So Python has a special data type called:

```
NoneType
```

And it has **only one value → None**.

---

# 3️⃣ Why Python Uses `None`

Python uses `None` when a function **does not return anything**.

Example:

```python
def fun():
    pass

print(fun())
```

Output

```
None
```

Because the function **returned nothing**.

---

# 4️⃣ Methods That Return `None`

Many Python methods modify objects **in-place** and return `None`.

Example:

```python
l = [3,1,2]
print(l.sort())
```

Output

```
None
```

Because:

```
sort() modifies the list directly
```

After sorting:

```python
print(l)
```

Output

```
[1,2,3]
```

---

# 5️⃣ Checking for `None`

Best practice is to use **`is` operator**.

Correct way:

```python
x = None

if x is None:
    print("No value")
```

Wrong way (not recommended):

```python
if x == None:
```

Because `None` is a **singleton object**.
