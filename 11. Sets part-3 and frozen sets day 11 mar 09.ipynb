## **Set inside a Set in Python**

### 1️⃣ Definition

In Python, a **set cannot contain another set** as an element.

Reason:
A **set is mutable**, and Python **sets only allow immutable (hashable) elements**.

So this is **not allowed** ❌

```python
s1 = {1, 2, 3}
s2 = {4, 5}

s = {s1, s2}   # Error
```

Output:

```
TypeError: unhashable type: 'set'
```

---

# 2️⃣ Why Python does NOT allow this

Python sets are implemented using **hash tables**.

Every element inside a set must be:

✔ **Hashable**
✔ **Immutable**

But a **set is mutable**, meaning it can change.

Example:

```python
s = {1,2,3}
s.add(4)
```

Because it can change, Python **cannot calculate a fixed hash value**, so it **cannot store it inside another set**.

---

# 3️⃣ The Correct Way → Use `frozenset`

Python provides **`frozenset`**, which is an **immutable version of set**.

Since it is immutable, it **can be stored inside another set**.

### Example

```python
s1 = frozenset({1,2})
s2 = frozenset({3,4})

s = {s1, s2}

print(s)
```

Output

```
{frozenset({1, 2}), frozenset({3, 4})}
```

Now this works because **frozenset is immutable**.

---

## **frozenset**

* **'frozenset'** is one of the pre-defined class and treated as set data type.
* The purpose of frozenset data type is that **"To store Multiple Values either Similar Type or Different Type or Both the Types in
Single Object with Unique Values"**.
* The elements of frozenset must be obtained from different objects like set, tuple and list.
* **Syntax:** `frozensetobj = frozenset(set/list/tuple)`


* An Object of frozenset never maintains **Insertion Order** bcoz PVM can display any one of the possibility of elements of frozenset object.
* On the object of frozenset, we can't perform **Indexing and Slicing Operations** bcoz frozenset object never maintains Insertion Order.
* An object of frozenset belongs to **Immutable** bcoz frozenset object does not support item assignment and not possible to modify / Change / add.
* We can create two types of frozenset objects. They are:
* **a) Empty frozenset**
* **b) Non-Empty frozenset**



---

### **a) Empty frozenset:**

* An Empty frozenset is one, which does not contain any elements and whose length is 0.

---
 Syntax: s1=frozenset()


### **b) Non-Empty frozenset:**

* A Non-Empty frozenset is one, which contains elements and whose length is >0.
* **Syntax:**
* `frozensetobj=frozenset({ val1, val2, ....val-n })`
* `frozensetobj=frozenset(( val1, val2, ....val-n ))`
* `frozensetobj=frozenset([ val1, val2, ....val-n ])`



---

### **NOTE:**

The Functionality of frozenset is exactly similar to set. But an object of set belongs to both **Mutable** (add, remove, pop, discard...etc)
and also **Immutable** in the case of Item assignment. Whereas frozenset object belongs to **Immutable** bcoz neither Possible to perform 
    add, remove, pop, discard...etc nor possible to perform Item assignment.

## **Why Tuples Cannot Be Copied in Python**

### 1️⃣ First Important Truth

Actually, **tuples can be copied**, but **copying a tuple is usually unnecessary**.

Reason: **tuples are immutable** (cannot be changed).

Because they cannot change, Python **often returns the same object instead of creating a new copy**.

---

# 2️⃣ What is a Tuple Copy?

Copying means **creating another object with the same elements**.

Example with lists:

```python
l1 = [1,2,3]
l2 = l1.copy()

print(l1 is l2)
```

Output

```
False
```

Two **different objects** are created.

---

# 3️⃣ Tuple "Copy"

Example:

```python
t1 = (1,2,3)
t2 = tuple(t1)

print(t1 is t2)
```

Output

```
True
```

Python returns the **same object**, not a new copy.

Reason: since tuples **cannot change**, copying is unnecessary.

---

# 4️⃣ Memory Optimization (Important Concept)

Python **optimizes memory**.

If an object is **immutable**, Python may reuse the same object.

Example:

```python
t1 = (1,2,3)
t2 = t1

print(id(t1))
print(id(t2))
```

Both **IDs are same**.

This saves **memory and improves speed**.

---

Python does **not need multiple copies**, because the tuple **cannot change**.
