]

## `update()` in Python Sets

### Core Idea

`update()` **adds elements from another iterable to the set**.

* It **modifies the original set** (in-place).
* It **does not return a new set**.
* It **automatically removes duplicates** (because sets only keep unique elements).

---

# Syntax

```python
set1.update(iterable)
```

`iterable` can be:

* set
* list
* tuple
* string
* any iterable object

---

# Basic Example

```python
A = {1, 2, 3}
B = {3, 4, 5}

A.update(B)

print(A)
```

Output

```
{1, 2, 3, 4, 5}
```

Explanation:

* Elements of **B** are added to **A**
* Duplicate `3` appears only once.

---

# Important Concept

`update()` is similar to **union**, but there is one key difference:

| Method     | Result                    |
| ---------- | ------------------------- |
| `union()`  | creates **new set**       |
| `update()` | **modifies existing set** |

Example:

```python
A = {1,2,3}
B = {4,5}

C = A.union(B)

print(A)   # unchanged
print(C)   # new set
```

But:

```python
A = {1,2,3}
B = {4,5}

A.update(B)

print(A)   # modified
```

---

| Operation                | Python Method             | Operator | Description                                       |                                     |
| ------------------------ | ------------------------- | -------- | ------------------------------------------------- | ----------------------------------- |
| **Union**                | `.union()`                | `        | `                                                 | All unique elements from both sets. |
| **Intersection**         | `.intersection()`         | `&`      | Only elements present in **both** sets.           |                                     |
| **Difference**           | `.difference()`           | `-`      | Elements in the **first set but not the second**. |                                     |
| **Symmetric Difference** | `.symmetric_difference()` | `^`      | Elements in **either set but not in both**.       |                                     |

# Example to Understand All Together

```python
A = {1,2,3,4}

````python
# Initializing the sets
cricket_players = {"Rohit", "Kohli", "Rossum"}
tennis_players = {"Travis", "Kinney", "Rossum"}

# a) Names of all players playing all types of games (Union)
# Note: In set terminology, "all players" usually refers to the Union.
all_players = cricket_players.union(tennis_players)
print("All players are ",(all_players))
````

````python
# b) Players playing BOTH Cricket and Tennis (Intersection)
both_games = cricket_players.intersection(tennis_players)
print(" Playing both games are :", both_games)
````

````python
# c) Only Cricket but not Tennis (Difference)
only_cricket = cricket_players.difference(tennis_players)
print(" Only Cricket is :", only_cricket)
````

````python
# d) Only Tennis but not Cricket (Difference)
only_tennis = tennis_players.difference(cricket_players)
print(" Only Tennis:", only_tennis)
````

# e) Exclusively playing Cricket and Tennis (Symmetric Difference)
# This finds players who play one or the other, but NOT both.

````python
exclusive_players = cricket_players.symmetric_difference(tennis_players)
print(" Exclusively playing one game:", exclusive_players)
````
