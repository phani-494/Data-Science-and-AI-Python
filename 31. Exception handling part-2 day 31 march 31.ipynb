### 1. Simple Definition
**Pre-Defined Exceptions** are error-reporting objects automatically provided by the Python interpreter. They act as "red flags" that the system raises when it encounters a standard logical or mathematical error during code execution.

---

### 2. Core Theory
In Python, all exceptions are **classes**. When an error occurs, Python "instantiates" (creates) an object of that specific error class. All built-in exceptions are organized in a hierarchy, descending from the base class `BaseException`.

The most common exceptions actually derive from the `Exception` class. This hierarchy allows you to catch specific errors (like `ValueError`) or broad categories (like `ArithmeticError`).



---

### 3. Examples (Basic → Advanced)

**Basic: Division by Zero**
```python
# System raises ZeroDivisionError
result = 10 / 0 
```

**Intermediate: Improper Data Types**
```python
# System raises TypeError because you can't add string to int
age = 25 + " years" 
```

**Advanced: Sequence & Mapping Errors**
```python
# IndexError: Accessing outside list bounds
my_list = [1, 2, 3]
print(my_list[5]) 

# KeyError: Accessing a non-existent dictionary key
my_dict = {"name": "Gemini"}
print(my_dict["age"]) 
```

---

### 4. Internal Working
1. **Detection:** The Python Interpreter detects an operation that violates language rules (e.g., $x / 0$).
2. **Object Creation:** It stops execution and creates an **Exception Object** containing the error type, the message, and the traceback (where it happened).
3. **Propagation:** Python looks for a `try-except` block. If none is found, it propagates the error up to the main program.
4. **Termination:** If still unhandled, Python prints the Traceback to the console and crashes the program.

---

### 5. Real-world Use Cases
* **Data Validation:** Using `ValueError` to catch cases where a user enters text in a "Phone Number" field.
* **File Handling:** Using `FileNotFoundError` to suggest a default configuration file if the user's specific file is missing.
* **API Integration:** Using `KeyError` to handle cases where a JSON response from a server is missing an expected data field.

---

### 6. Common Mistakes (VERY Important)
* **Catching 'Exception' blindly:** * *Wrong:* `except Exception:` (This hides bugs you didn't intend to catch, like `KeyboardInterrupt`).
    * *Right:* Catch the specific error: `except ValueError:`.
* **Confusing SyntaxError with Exceptions:** A `SyntaxError` happens *before* the code runs (parsing phase). Exceptions only happen *while* the code is running (execution phase).
* **Ignoring the Traceback:** Many beginners look only at the error name. The **line number** in the traceback is your best friend for debugging.

---

### 7. Interview Insights
* **Q: What is the difference between `IndexError` and `KeyError`?**
    * *A:* `IndexError` occurs in sequences (lists, tuples) when an integer index is out of range. `KeyError` occurs in mappings (dictionaries) when a specific key is not found.
* **Q: Can you handle a `SystemExit` exception?**
    * *A:* Yes, but you shouldn't. It's a sub-class of `BaseException` (not `Exception`) designed to shut down the script cleanly.
* **Q: What is the 'MRO' of exceptions?**
    * *A:* It stands for Method Resolution Order. Python searches the exception hierarchy from the most specific (child) to the most general (parent).

**Programmer/User/Custom Defined Exceptions** in Python:

---

## 1. Simple Definition
**User-Defined Exceptions** are specialized error classes created by a programmer to handle specific business logic failures that Python's built-in exceptions (like `ValueError` or `TypeError`) don't covers.

---

## 2. Core Theory (Clean + Deep)
In Python, creating a custom exception is done by **Inheritance**. You define a new class that inherits from the built-in `Exception` class (or one of its subclasses).

* **Inheritance Hierarchy:** Your custom class becomes part of the Python Exception tree.
* **The `raise` Keyword:** Unlike built-in errors that Python triggers automatically, you must manually trigger custom exceptions using the `raise` keyword when a specific condition is met.
* **Constructor (`__init__`):** You can override the constructor to accept custom arguments (like an error code or a specific message).



---

## 3. Examples (Basic → Advanced)

### **Basic: Simple Custom Error**
```python
class InvalidAgeError(Exception):
    """Raised when the input age is negative"""
    pass

age = -5
if age < 0:
    raise InvalidAgeError("Age cannot be negative!")
```

### **Advanced: Exception with Data Payload**
You can store extra information inside the exception object to help the "catch" block make better decisions.
```python
class InsufficientFundsError(Exception):
    def __init__(self, balance, amount):
        self.balance = balance
        self.amount = amount
        self.shortfall = amount - balance
        super().__init__(f"Need {self.shortfall} more to complete transaction.")

# Usage
try:
    current_bal = 500
    withdraw = 1000
    if withdraw > current_bal:
        raise InsufficientFundsError(current_bal, withdraw)
except InsufficientFundsError as e:
    print(f"Error: {e} | Current Balance: {e.balance}")
```

---

## 4. Internal Working
1.  **Class Definition:** Python registers your class as an exception type.
2.  **Condition Check:** Your code evaluates a business rule (e.g., `if balance < withdrawal`).
3.  **Instantiation & Raising:** When `raise MyError()` is called, Python halts the regular flow and creates an instance of your class.
4.  **Stack Search:** Python searches for a matching `except MyError:` block. If found, it passes the instance to that block; otherwise, it crashes with your custom message.

---

## 5. Real-world Use Cases
* **Banking Apps:** Raising `InsufficientBalanceError` during a withdrawal.
* **Authentication:** Raising `AccountLockedError` after 3 failed login attempts.
* **E-commerce:** Raising `OutOfStockError` when a user adds a sold-out item to their cart.
* **Data Validation:** Raising `InvalidGSTNumberError` if a tax ID doesn't match the required format.

---

## 6. Common Mistakes (VERY Important)
* **Not Inheriting from `Exception`:** If you inherit from `BaseException` instead of `Exception`, your error might bypass standard `except Exception:` blocks, making it hard to catch.
* **Over-complicating:** Don't create a custom exception if a built-in one works. Use `ValueError` for bad inputs unless you need specific metadata.
* **Empty `pass` classes:** While `class MyError(Exception): pass` is valid, it’s a missed opportunity to add helpful debugging attributes (like the "shortfall" example above).

---

## 7. Interview Insights
* **Q: Why inherit from `Exception` instead of `BaseException`?**
    * *A:* `BaseException` includes system-level events like `SystemExit` and `KeyboardInterrupt`. If you inherit from it, a general `except Exception:` won't catch your custom error, which can lead to unexpected crashes.
* **Q: Can a custom exception have multiple catch blocks?**
    * *A:* Yes, you can catch the specific custom error first, and then have a fallback `except Exception:` block for other unforeseen issues.
* **Q: What is the benefit of custom exceptions in a large team?**
    * *A:* It improves **Readability**. Seeing `InvalidTransactionError` in a log file is much more helpful than seeing a generic `RuntimeError`.


# 🧠 1. What is `try-except-else`?

### ✅ Simple Definition

* `try` → code that may cause error
* `except` → handles error
* `else` → runs **only if NO exception occurs**

---

# ⚙️ Basic Structure

```python
try:
    # risky code

except SomeError:
    # handle error

else:
    # runs only if try succeeds
```

---

# 🔥 Core Rule (VERY IMPORTANT)

👉 `else` executes **ONLY when try block has NO error**

---

# 🧪 Example 1 (Simple)

```python
try:
    n = int(input("Enter number: "))
except ValueError:
    print("Invalid input")
else:
    print("You entered:", n)
```

---

### 🔍 Case 1: Input = `abc`

* `int("abc")` → ValueError
* goes to `except`
* ❌ `else` NOT executed

---

### 🔍 Case 2: Input = `10`

* No error
* `except` skipped
* ✅ `else` runs

---

# ⚙️ Why `else` Exists (Important Insight)

You might think:
👉 “Why not just write everything in try?”

Bad practice ❌

---

### ❌ Wrong Way

```python

    n = int(input())
    print(n)   # this should not be inside try
except:
    print("error")
```

👉 Problem:

* If `print()` fails (rare but possible), it gets caught wrongly

---

### ✅ Correct Way

```python
try:
    n = int(input())
except ValueError:
    print("Invalid input")
else:
    print(n)
```

👉 Clean separation:

* `try` → only risky code
* `else` → safe code

# 🔬 Internal Flow

### Step-by-step:

1. Python enters `try`
2. If error occurs → jumps to `except`
3. If NO error → skips `except`
4. Executes `else`

Exception handling in Python (and many other languages) is a way to manage errors gracefully so your program doesn't just crash when something goes wrong. Think of it like a safety net for your code.

Here is the breakdown of how the **try**, **except**, and **else** blocks work together:

---

### 1. The `try` Block (The "Experiment" Block)
The `try` block is where you put the code that might cause an error (an exception). You are essentially telling the computer, "Try to run this, but keep an eye out for trouble."
* **Purpose:** To monitor for exceptions.
* **Behavior:** If an error occurs, the rest of the code inside the `try` block is skipped immediately.

### 2. The `except` Block (The "Safety Net" Block)
The `except` block only runs **if an error occurred** inside the `try` block. It prevents the program from crashing by "catching" the error and allowing you to handle it (e.g., printing a user-friendly message).
* **Purpose:** To process and handle the error.
* **Behavior:** If no error occurs in the `try` block, this section is completely ignored.

### 3. The `else` Block (The "Success" Block)
The `else` block is a bit unique. It only runs **if no exceptions were raised** in the `try` block. It is the perfect place for code that should only execute if the "try" section was 100% successful.
* **Purpose:** To separate the "dangerous" code (in `try`) from the logic that follows a successful run.
* **Behavior:** If an exception is caught by the `except` block, the `else` block will **not** run.



---

### A Simple Example
Imagine you are asking a user to divide two numbers.

```python
try:
    num1 = int(input("Enter a number: "))
    num2 = int(input("Enter another number: "))
    result = num1 / num2
    
except ZeroDivisionError:
    # This runs ONLY if the user enters 0 as the second number
    print("Error: You cannot divide by zero!")

except ValueError:
    # This runs ONLY if the user enters text instead of a number
    print("Error: Please enter valid numbers only.")

else:
    # This runs ONLY if everything above worked perfectly
    print(f"Success! The answer is {result}")
```

### Summary Table

| Block | When does it run? |
| :--- | :--- |
| **try** | Always. |
| **except** | Only if an error (exception) occurs in the `try` block. |
| **else** | Only if **no** error occurs in the `try` block. |
