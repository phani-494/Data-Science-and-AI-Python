## **Simple Explanation**

### **What is an Exception?**
In Python, an **Exception** is just a fancy word for an error that happens while a program is running. Instead of the whole program crashing blindly, Python "raises" an exception.

### **The Hierarchy (The "Family Tree")**
Think of exceptions like a family tree:
* **The Ancestor (`BaseException`):** Everything starts here. 
* **The Parents (`Exception`):** Most errors you deal with are children of the `Exception` class.
* **The Children (`ValueError`, `TypeError`, etc.):** These are specific types of mistakes. For example, if you try to divide by zero, Python looks for the `ZeroDivisionError` "child."

### **Inheritance: "Is-A" Relationship**
**Inheritance**. This is a way to make a new class that takes features from an old one.
* **Parent/Teacher:** The original class.
* **Child/Student:** The new class that "inherits" everything from the parent. 
* **Syntax:** When you see `class Student(Teacher):`, it means "The Student is a type of Teacher" (it gets all the Teacher's properties).

### **Custom Exceptions**
This section shows how you can make your own errors. If you are building a banking app, Python doesn't have a built-in "Wrong PIN" error. 

You create it by making a child of the `Exception` class:

`class PinError(Exception): pass`

By doing this, your custom error acts just like the official Python errors!

## ✅ 2. Create `bank.py`

```python
from exceptions import InsufficientBalanceError, InvalidAmountError

balance = 5000

def withdraw(amount):
    global balance

    if amount <= 0:
        raise InvalidAmountError("Amount must be greater than 0")

    if amount > balance:
        raise InsufficientBalanceError("Not enough balance")

    balance -= amount
    return balance
```

👉 Business logic + raising exceptions

---

## ✅ 3. Create `main.py`

```python
from bank import withdraw
from exceptions import InsufficientBalanceError, InvalidAmountError

try:
    amt = int(input("Enter amount: "))
    bal = withdraw(amt)
    print("Remaining Balance:", bal)

except InvalidAmountError as e:
    print("Invalid Input:", e)

except InsufficientBalanceError as e:
    print("Transaction Failed:", e)
```
