# Local and global variables 

## Local variables 

**=>** **Local variables** are used in Function Body.

**=>** The Purpose of Local Variables is that "To store Temporary Results after Function Processing".

**=>** Local Variables can be accessed inside of Corresponding function definition only but not possible to access other part of the program.

## Global Variables:

-------------------------------------------------------------------------
**=>** **Global Variables** are those which are used for Providing Common Value for all Different Function Definitions.

**=>** Global Variables to be defined before all the function calls. So that we can access global variable values in all those Function definitions.

**=> Syntax:**
-----------
```python
# Global Variable Definition Area
# -------------------------------

def fun1():
    --------------------
    --------------------

def fun2():
    --------------------
    --------------------

# ... (Additional functions)

def fun-n():
    --------------------
    --------------------
```

---

### Understanding Scope

To visualize this:
* **Local Variables** live inside a "box" (the function). Once the function finishes running, the box is destroyed, and the variable is gone.
* **Global Variables** live in the open "environment" outside the boxes, so every function can look out and see them.


````python
# here sub and lang are the local variables and can be accessed only inside the fuunction
def learnAI():
    sub="AI"
    lang="Python"
    print("To develop {} Apps we need to use {}".format(sub,lang))
learnAI()
````



```python
# Main Program
Var1=Val1

------------------
Var_n=Val_n # Here Var1,Var2,......Var-n are called Global Variables
fun1()
fun2()
-------
fun-n()
```

**=>** Hence **Var1,Var2,......Var-n** are the **Global Variables** defined before Function Definitions, so that We can access those values inside of Function Definitions.

---

### Summary of Scope Rules
Based on the notes you've shared so far, here is a quick cheat sheet for your studies:

| Feature | Local Variable | Global Variable |
| :--- | :--- | :--- |
| **Declaration** | Inside a function body. | Outside all functions (Main Program). |
| **Lifetime** | Created when function starts; destroyed when it ends. | Persists throughout the program execution. |
| **Accessibility** | Only within that specific function. | Accessible by any function in the program. |
| **Purpose** | Temporary processing/storage. | Shared data or common values across functions. |


# 📘 `global` Keyword in Python (Complete Mastery)

---

## 🔹 1. What is `global`?

👉 `global` is a keyword used **inside a function** to tell Python:

> “This variable belongs to the global scope, not local.”

---

## 🔹 2. Why Do We Need `global`?

Because of this rule:

> If you assign a variable inside a function → Python treats it as **local by default**

---

## 🔥 3. The Problem Without `global`

```python
x = 10

def func():
    x = x + 1   # ❌ error

func()
```

### ❌ Error:

```
UnboundLocalError
```

---

## 🧠 What Python Thinks

Inside `func()`:

```python
x = x + 1
```

Python thinks:

* “x is local (because of assignment)”
* But local `x` has no value yet ❌

---

## 🔹 4. Using `global` (Fix)

```python
x = 10

def func():
    global x
    x = x + 1

func()
print(x)
```

### ✅ Output:

```
11
```

---

## 🧠 What Happens Internally

* `global x` → tells Python:
  👉 “Use the global variable x”
* So no local variable is created

---

## 🔹 5. Multiple Variables

```python
a = 5
b = 10

def update():
    global a, b
    a = a + 1
    b = b + 2

update()
print(a, b)
```

---

## 🔹 6. Creating Global Variable Inside Function

```python

    global x
    x = 100

func()
print(x)
```

👉 Variable created globally from inside function

## 🔹 2. Local Variable (DEEP)


### 🧠 What actually happens:

* After function ends → that memory is **destroyed**

👉 So:

* Local variables are **temporary**
* They **do not exist outside**

---

### ❌ Error Example


👉 Because `x` is destroyed after function ends

👉 Variable defined **outside all functions**

```python

def func():
    print(x)

func()
```

### 🧠 What happens:

* `x` is stored in **global memory**
* Any function can **access (read)** it

## **global key word**

==================================================

**=> When we want MODIFY the GLOBAL VARIABLE values in side of function defintion then global variable names must be preceded with 'global' keyword otherwise we get "UnboundLocalError: local variable names referenced before assignment"**

### **Syntax:**
------------------
```python

    -----------------------

    -----------------------
    global var1, var2...var-n
        # Modify var1, var2....var-n
    -----------------------
```

**NOTE:** To **MODIFY** Global variable Values inside of Function Definition, we use **global Keyword** (Mandatory to write).

**NOTE:** To **ACCESS** Global variable Values inside of Function Definition, we **Don't use global Keyword**.

==================================================


---

### **Key Takeaway**
The main distinction your notes are making is between **reading** and **writing**:

| Action | Keyword Required? | Reason |
| :--- | :--- | :--- |
| **Reading (Accessing)** | **No** | Python searches for the name in the local scope, then automatically looks in the global scope. |
| **Writing (Modifying)** | **Yes** | Without `global`, Python assumes you are creating a *new* local variable with the same name. |

# 📘 `nonlocal` in Python (Deep & Clear)

## 🔹 1. What is `nonlocal`?

👉 `nonlocal` is used **inside nested functions** (function inside another function)

It tells Python:

> “This variable is NOT local to this function—use the variable from the nearest outer function.”

---

## 🔥 2. Why Do We Need `nonlocal`?

Because of Python’s rule:

> If you assign a variable inside a function → it becomes **local**

---

## ❌ Problem Without `nonlocal`

```python
def outer():
    x = 10

    def inner():
        x = x + 1   # ❌ error

    inner()

outer()
```

---

### ⚠️ Error:

```
UnboundLocalError
```

---

## 🧠 Why Error Happens?

Inside `inner()`:

```python
x = x + 1
```

Python thinks:

* “x is local to inner()”
* But it is used before assignment ❌

---

# ✅ 3. Solution: `nonlocal`

```python
def outer():
    x = 10

    def inner():
        nonlocal x
        x = x + 1

    inner()
    print(x)

outer()
```

### ✅ Output:

```

```

## 🧠 What `nonlocal` Does

👉 It tells Python:

* Don’t create local variable
* Use variable from **outer function**



## Global and Local Variables and `globals()`

**=>** When we come across the same **Global Variable names** and **Local Variable names** in the same function definition, the **PVM (Python Virtual Machine)** gives preference to local variables but not for global variables.

**=>** In this context, to extract/retrieve the values of global variable names along with local variables, we must use **`globals()`**. 

It returns an object of `<class 'dict'>`, and this dict object stores all global variable names as **Keys** and global variable values as **Values**.

---


## Syntax for Extracting Global Variables

```python
var1 = val1
var2 = val2
------------------
var-n = val-n  # var1, var2...var-n are called global Variables

def functionname():
    ------------------
    var1 = val11
    var2 = val22
    ------------------
    var-n = val-nn # var1, var2...var-n are called local Variables
    
    # Extract the global variables values
    dictobj = globals()

    globalval1 = dictobj['var1'] # or dictobj.get("var1") or globals()['var1'] or globals().get('var1')
    globalval2 = dictobj['var2'] # or dictobj.get("var2") or globals()['var2']
    --------------------------------------------------
    --------------------------------------------------
```




### Key Takeaways from this logic:

* **`dictobj = globals()`**: This captures the entire global symbol table into a dictionary variable.
* **Accessing Values**: You have four primary ways to get a global value when a local variable has the same name:
    1.  **Direct Key Access:** `dictobj['var1']` (Raises error if key doesn't exist).
    2.  **Dictionary `.get()`:** `dictobj.get("var1")` (Returns `None` if key doesn't exist).
    3.  **Direct Function Call:** `globals()['var1']` (No need to create a temporary `dictobj`).
    4.  **Direct Function `.get()`:** `globals().get('var1')`.
