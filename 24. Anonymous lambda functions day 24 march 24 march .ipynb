# lambda or  anonymous functions 


## Anonymous Functions OR Lambda Functions

* **Definition:** Anonymous Functions are those which do not have any Function Name explicitly.
* **Purpose:** The purpose of an Anonymous Function is "To Perform Instant Operations."
* **Instant Operations:** These are operations used at that specific point in time only, with no interest in using them in other parts of the application or project.
* **The 'lambda' Keyword:** To define Anonymous Functions, we use the `lambda` keyword; hence, Anonymous Functions are called Lambda Functions.
* **Single Statement:** Since Anonymous Functions perform instant operations, they contain a **single executable statement only** (not multiple).
* **Implicit Return:** Anonymous Functions return the result automatically or implicitly (there is no need to use a `return` statement).

### Syntax for Defining Anonymous Function

```python
varname=lambda arguments : expression
```
--> arguments is simply parameters -list 



## Explanation (Continued)

* **`varname`**: Represents an object of `<class 'function'>` and it can be treated indirectly as a Function Call.
* **`lambda`**: Represents a Keyword and it is used for Defining Anonymous OR Lambda Functions.
* **`Params-List`**: Represents a List for Formal Parameters which are used for Holding OR Storing the Inputs coming from Function calls.
* **`Expression`**: Represents a Single Executable Statement and it is the Solution OR Logic for Instant Operations.
* **Implicit Return**: Anonymous Functions return the Result Automatically / Implicitly by executing a Single Executable Statement. There is no need to use a `return` statement.

### Question: Define a Function for adding Two Numbers

#### 1. By using Normal Functions
To do this with a standard named function, you would use the `def` keyword:

```python
def add(a, b):
    return a + b

# Calling the function
print(add(10, 20)) # Output: 30
```

#### 2. By using Anonymous Functions
Using the lambda syntax for a more concise, "one-liner" approach:

```python
add = lambda a, b : a + b

# Calling the function
print(add(10, 20)) # Output: 30
```

# 🧠 1. What is a Lambda Function (REAL understanding)

A **lambda function** is a **small anonymous (nameless) function** used for **short, one-time operations**.

### Syntax:

```python
lambda arguments: expression
```

👉 Key points:

* No `def`
* No function name
* Only **one expression** (no multiple statements)
* Automatically returns value

---

# ⚡ 2. Basic Example

```python
add = lambda a, b: a + b
print(add(2, 3))  # 5
```

👉 Same as:

```python
def add(a, b):
    return a + b
```

---

# 🔥 3. Where Lambda is ACTUALLY used (Important)

Most beginners misuse lambda. Real usage is here 👇

---

## ✅ (1) With `map()` → Data Transformation

### Real-world: Convert list of prices to GST added prices

```python
prices = [100, 200, 300]

gst_prices = list(map(lambda x: x * 1.18, prices))
print(gst_prices)
```

👉 Used in:

* Data preprocessing (AI/ML)
* Bulk transformations

---

## ✅ (2) With `filter()` → Selection

### Real-world: Filter passing students

```python
marks = [35, 80, 25, 90, 40]

passed = list(filter(lambda x: x >= 40, marks))
print(passed)
```

👉 Used in:

* Cleaning datasets
* Removing unwanted values

---

## ✅ (3) With `sorted()` → Custom Sorting (VERY IMPORTANT)

### Real-world: Sort students by marks

```python
students = [("A", 50), ("B", 80), ("C", 40)]

sorted_students = sorted(students, key=lambda x: x[1])
print(sorted_students)
```

👉 This is heavily used in:

* Data Science
* Backend systems
* Ranking problems

---

## ✅ (4) With `reduce()` → Aggregation

```python
from functools import reduce

nums = [1, 2, 3, 4]

product = reduce(lambda x, y: x * y, nums)
print(product)  # 24
```

👉 Used in:

* Mathematical computations
* Data aggregation

---

# ⚠️ 4. Important Limitations (Don’t ignore)

Lambda:

* ❌ Cannot have multiple statements
* ❌ Cannot use loops inside directly
* ❌ Hard to read if overused

---

# 🧠 5. When to Use vs NOT Use

## ✅ Use Lambda When:

* Function is **small (1 line)**
* Used **once**
* Inside `map`, `filter`, `sorted`

## ❌ Avoid Lambda When:

* Logic is complex
* Needs readability
* Used multiple times → use `def`

---

# Lambda Functions in Python

## What is it?
A lambda is a **small, anonymous (nameless) function** written in one line.

---

## Syntax
```python
lambda arguments : expression
```

---

## Regular Function vs Lambda

```python
# Regular function
def square(x):
    return x * x

# Same thing as lambda
square = lambda x: x * x

print(square(5))  # Output: 25
```

---

## Multiple Arguments

```python
add = lambda x, y: x + y
print(add(3, 4))  # Output: 7
```

---

## Where Lambdas Shine

### 1. `sorted()` — custom sorting
```python
students = [("Alice", 85), ("Bob", 72), ("Charlie", 90)]
sorted_students = sorted(students, key=lambda s: s[1])
# Sorted by marks: [('Bob', 72), ('Alice', 85), ('Charlie', 90)]
```

### 2. `map()` — apply to every item
```python
numbers = [1, 2, 3, 4]
squares = list(map(lambda x: x**2, numbers))
# Output: [1, 4, 9, 16]
```

### 3. `filter()` — keep matching items
```python
numbers = [1, 2, 3, 4, 5, 6]
evens = list(filter(lambda x: x % 2 == 0, numbers))
# Output: [2, 4, 6]
```

---

## Rules to Remember

| Rule | Detail |
|------|--------|
| Single expression only | No multi-line logic |
| Auto returns | No need to write `return` |
| Anonymous | No name needed (but can be assigned) |
| Keep it simple | For complex logic, use `def` |

---

## When to Use vs Avoid

✅ **Use lambda** → short, simple, one-time-use functions

❌ **Avoid lambda** → complex logic, reusable functions, anything needing readability
