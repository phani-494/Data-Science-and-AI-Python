### Purpose of Looping in Python 🔁

A **loop** is used to **execute a block of code repeatedly** until a condition is satisfied or a sequence is finished.

Instead of writing the same code many times, we use **loops to automate repetition**.

---

## 1️⃣ Why Looping is Needed

### 1. Avoid Rewriting Code

Without loops, you would have to repeat the same statement many times.

❌ Without loop

```python
print("Hello")
print("Hello")
print("Hello")
print("Hello")
print("Hello")
```

✅ With loop

```python
for i in range(5):
    print("Hello")
```

Loop prints **Hello 5 times automatically**.

# 1️⃣ Concept (What is a `while` Loop?)

A **`while` loop** repeatedly executes a block of code **as long as the given condition is `True`**.

The loop stops when the **condition becomes `False`**.

👉 It is called a **condition-controlled loop**.

Example idea:

```
while condition is True:
    keep executing
```

---

# 2️⃣ Syntax

```python
while condition:
    statements
```

Optional `else`:

```python
while condition:
    statements
else:
    statements
```

`else` runs **when loop ends normally (without break)**.

---

# 3️⃣ Flow of Execution

```
1 Check condition
2 If True → execute loop body
3 Go back and check condition again
4 If False → exit loop
```

---

### **Syntax for `while` and `while..else` Loops**

---

#### **1. Standard `while` Loop Syntax**

```python
while(Test Cond):
    ---> Statement-1   | Indentation
    ---> Statement-2   | Block of statements
    ---> Statement-n   |
Other statements in Prog

```

**OR**

#### **2. `while..else` Loop Syntax**

```python
while(Test Cond):
    ---> Statement-1   | Indentation
    ---> Statement-2   | Block of statements
    ---> Statement-n   |
else:
    ---> Else Block of Stmts
Other Stmts in Program

```

---

### **Explanation:**

* **Keywords:** Here `while` and `else` are called **Keywords**.
* **Test Condition:** The `Test Cond` can result in either **True** or **False**.
* **Execution Flow (True):** If the `Test Cond` is **True**, the Python Virtual Machine (PVM) executes the **Indentation Block of statements**. Once executed, the PVM control returns to the `Test Cond`. This process repeats for a finite number of times until the `Test Cond` becomes **False**.
* **Execution Flow (False):** Once the `Test Cond` becomes **False**, the PVM executes the **else block of statements** (if present). After that, it proceeds to execute any **Other Statements** written in the program.

### **Key Logic Note**

In Python, the `else` block in a loop is unique: it executes **only if the loop finishes naturally** (i.e., the condition becomes false). If the loop is terminated prematurely by a `break` statement, the `else` block is skipped entirely.

````python
n=int(input("enter a number:"))
print("*"*30)
print("Even numbers are :")
print("*"*30)
if (n%2==0 and n>0):
    i=n
    while(i>=1): # even writing i>=2 also fine here 
        print(i)
        i=i-2
    else:
        print("*"*30)
elif (n%2!=0 and n>0):
    i=n-1
    while(i>=1):   # even writing i>=2 also fine here 
        print(i)
        i=i-2
    else:
        print("*"*30)
else:
    print("Invalid input")
````

### Why We Use `else` in a `while` Loop in Python 🔁

In Python, the **`else` block in a `while` loop** executes **only when the loop finishes normally** (when the condition becomes `False`).
It **does not execute if the loop stops because of a `break` statement**.

So the purpose of `else` in a `while` loop is to **run some code after the loop completes successfully**.

---

## 1️⃣ Basic Syntax

```python

    # loop code
else:
    # runs when loop ends normally
```

# sum of digits of a given  number 

````python
n=int(input("Enter a number:"))
sum=0
t=n
while(n!=0):
    rem=n%10 # to get the last digit
    sum=sum+rem
    n=n//10
else:
    print("Sum({})={}".format(t,sum))
````

````python
n=input("Enter a number:")
sum=0
t=n
for i in n:
    sum=sum+int(i)
else:
    print("Sum({})={}".format(t,sum))
````

````python
# when we don't spaces as special symbols and don't want to count them, then we can use the split function
s1=input("enter the word:")
words=s1.split()
print(words)
count=0
for word in words:
    for ch in word:
        if (not ch.isalnum()):
            print(ch)
            count=count+1
print(count)
````

# 5️⃣ How It Works

### `eval()` Function

`eval()` converts input into its **actual data type**.

Example:

| Input     | Stored As |
| --------- | --------- |
| `10`      | integer   |
| `"hello"` | string    |
| `3.5`     | float     |
| `True`    | boolean   |

Without `eval()` → everything becomes **string**.

# program to take list of values and display them in both sorting orders 

sort() returns None

The list.sort() method modifies the list in place and does not return a new list.

````python
n=int(input("Enter the no of values user wants: "))
l1=[]
for i in range(n):
    l1.append(float(input("Enter {} value :".format(i+1))))
print(l1)
l1.sort()
print(l1)
print(l1[::-1])
````

| Function       | Behavior                        |
| -------------- | ------------------------------- |
| `list.sort()`  | modifies list, returns **None** |
| `sorted(list)` | returns **new sorted list**     |


Concept

sorted() is a built-in Python function used to sort any iterable (list, tuple, string, set, dictionary).

Important point:

✔ It does NOT modify the original object
✔ It returns a new sorted list

````python
n = int(input("Enter number of values: "))
l1 = [float(input(f"Enter {i+1} value: ")) for i in range(n)]

print("Original:", l1)
print("Ascending:", sorted(l1))
print("Descending:", sorted(l1, reverse=True))
````

# Why You Cannot Enter a String with `eval(input())`

When you write:

```python
x = eval(input("Enter the value: "))
```

Python does **two steps**:

### Step 1 → `input()`

Takes user input **as a string**.

Example input:

```
hello
```

Python receives:

```python
"hello"
```

---

### Step 2 → `eval()`

`eval()` **evaluates the input as a Python expression**.

So Python tries to interpret:

```python
hello
```

as a **variable name**, not a string.

Since `hello` is **not defined**, Python gives an error:

```
NameError: name 'hello' is not defined
```

---

# How to Enter a String with `eval()`

You must use **quotes**.

Example input:

```
"hello"
```

or

```
'hello'
```

Now Python evaluates it as a **string**.

Example program:

```python
x = eval(input("Enter value: "))
print(x)
print(type(x))
```

Input

```
"Python"
```

Output

```
Python
<class 'str'>
```

It automatically converts input into **actual Python data types**.

```python

```

Input:

```

```


# ⚠️ Warning (Important)

`eval()` is **dangerous in real applications** because users can execute code.

Example:

```
__import__("os").system("shutdown")
```

So professionals often avoid `eval()`.

---

# Safer Alternative

Use `input()` and convert manually.

Example:

```python
x = input("Enter string: ")
```

or

```python
x = int(input("Enter number: "))
```
