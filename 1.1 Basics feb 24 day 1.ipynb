# In Python, a library is a collection of pre-written code (modules and packages) that helps you perform common tasks without writing everything from scratch.

Simple definition

 1. A Python library = ready-made tools you can reuse
 

## Why Python libraries are important

Save time

Reduce errors

Make code shorter & cleaner

Allow you to build powerful applications quickly

# Term	Meaning

Module: Single .py file

Package: Folder of modules

Library:  Collection of modules & packages

Think of a Python library as a pre-packaged toolkit for a specific job.


## How It Works;

A library contains a set of modules (files ending in .py) that include pre-defined functions, classes, and variables.

Installation: Most libraries are downloaded via a tool called pip.

Importing: You bring the "toolbox" into your script using the import command.

Usage: You call the specific tool you need.
