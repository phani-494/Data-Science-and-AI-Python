# Flow Control statements  or control structures

## **Flow Control Statements in Python**

**=>The purpose of Flow Control Statements in Python is that "To Perform Certain Operation(X-Operation in the case 
of True OR Y-Operation in the Case of False) Only Once OR To Perform Certain Operation Repeatedly for Finite Number of Times Until 
Test Condition becomes False".**

**=>In Python Programming, we have 3 Types of Flow Control Statements.**

1. **Conditional OR Selection OR Branching Statements**
2. **Looping OR Iterating OR Repetative Statements**
3. **Transfer Flow Statements**


---

**=>The purpose of Conditional OR Selection OR Branching Statements is that "To Perform X-Operation in the Case of

1. **Simple if statement**

## 1. if

    statement1

    .........

Key points ⚡

1. condition must return True or False

2. : (colon) is mandatory (colon act as start and end braces and the statements which are intended are prt of the if block here.

3. Indentation is mandatory in Python

4. All the statements  inside 'if' are  referred as the indendation block or suite.



## **1. Simple if statement**

### **Syntax:**

```python
if (Test Cond) :  # ':' is the Indentation Symbol
    ----> Statement-1  |
    ----> Statement-2  | Indentation Block (OR)
    ----> ...........  | Indentation Suite
    ----> Statement-n  |
-----------------------
Other Statements in Program

```

### **Explanation:**

* **=>** Here **'if'** is a Keyword.
* **=>** Here **Test Cond** Can be either **True** or **False**.
* **=>** If the **Test Cond** is **True**, then **PVM** (Python Virtual Machine) executes the **Indentation Block of Statements** and later executes **Other statements**.
* **=>** If the **Test Cond** is **False**, then **PVM** executes **Other statements** without executing the **Indentation Block of Statements**.

---

### **Flow Chart Logic:**

1. **Entry:** The program reaches the `if` statement.
2. **Test Cond?** The condition is evaluated.
* **If True:** It moves to **Execute Indentation Block of Statements**, then proceeds to **Execute Other Statement in Program**.
* **If False:** It skips the block and moves directly to **Execute Other Statement in Program**.

# 2. if..else statement

### Syntax:

```python
if (Test Cond) :
    -------Statement-1  | Indentation
    -------Statement-2  | Block-I
    -------Statement-n  |
else:
    -------Statement-11 | Indentation
    -------Statement-12 | Block-II
    -------Statement-1n |
Other Statements in Program

```

---

### Flow Chart:

* **Entry:** The program enters the decision phase.
* **Test Cond?:** A diamond decision box evaluates the condition.
* **True:** If the condition is met, the program flows to **Execute Indentation Block of Stmts-I**.
* **False:** If the condition is not met, the program flows to **Execute Indentation Block of Stmts-II**.


* **Merge:** After executing either block, both paths lead to **Execute Other Statements in Program**.

---

### Explanation:

* Here **if** and **else** are the keywords.
* If the **Test Cond** is **True**, then **PVM** (Python Virtual Machine) executes **Indentation Block of Statements-I** and later executes **Other Statements** in program.
* If the **Test Cond** is **False**, then **PVM** executes **Indentation Block of Statements-II** and later executes **Other Statements** in program.

---


````python
a=float(input("Enter a number for A"))
b=float(input("Enter a number for B"))
if (a>b):
    print("A is bigger as {}>{}".format(a,b))
else: # in this case when first if is True , then else part won't be executed, thus time complexity will be reduced here 
    if (b>a):
        print("B is bigger as as {}>{}".format(b,a))
    else:
        print("Both  equal as {}={}".format(a,b))
    
````

## 🔁 Flow Control in Python

**Flow control** means **controlling the order in which statements execute in a program**.

Normally Python executes code **top → bottom**, but flow control lets us **change the execution path based on conditions or repetition**.

📌 In real programs, flow control helps make **decisions, repeat tasks, and stop execution when needed**.

---

# 1️⃣ Conditional Statements (Decision Making)

These execute code **based on conditions**.

## 1. `if` statement

### Concept

Executes a block **only if the condition is True**.
The if statement is used for decision making in Python.

It allows the program to execute certain code only when a condition is True.

```python
age = 20

if age >= 18:
    print("You can vote")
```

### How it works

1. Python checks the condition.
2. If **True → code runs**
3. If **False → skipped**

### Real-world use 🌍

* Login validation
* ATM withdrawal check
* Checking eligibility

Example:

```python
balance = 5000
withdraw = 2000

if withdraw <= balance:
    print("Transaction successful")
```

---

## 2. `if-else`

### Concept

Used when there are **two possible outcomes**.

```python
num = 10

if num % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

### Real-world use 🌍

* Pass / Fail result
* Payment success / failure
* Access granted / denied

---

## 3. `if-elif-else`

### Concept

Used when **multiple conditions exist**.

```python
marks = 85

if marks >= 90:
    print("Grade A")
elif marks >= 70:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
```

### Real-world use 🌍

* Grading systems
* Salary tax calculation
* AI decision rules

---

# 2️⃣ Looping Statements (Repetition)

Loops allow **repeating code multiple times**.

Two types:

1️⃣ `for loop`
2️⃣ `while loop`

---

# 3️⃣ `for` Loop

### Concept

Used to **iterate over a sequence**.

Examples of sequences:

* list
* tuple
* string
* set
* range

Example:

```python
for i in range(5):
    print(i)
```

Output

```
0
1
2
3
4
```

### Real-world use 🌍

* Processing datasets
* Reading files
* Training AI models

Example:

```python
names = ["Rahul", "Ravi", "John"]

for name in names:
    print(name)
```

---

# 4️⃣ `while` Loop

### Concept

Runs **until the condition becomes False**.

```python
i = 1

while i <= 5:
    print(i)
    i += 1
```

### Real-world use 🌍

* Waiting for user input
* Running servers
* Game loops

Example:

```python
password = ""

while password != "python123":
    password = input("Enter password: ")
```

---

# 5️⃣ Loop Control Statements

These control loop execution.

---

## `break`

Stops the loop immediately.

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

Output

```
0 1 2 3 4
```

Use case 🌍

* Stop search when item found
* Exit infinite loop

---

## `continue`

Skips the current iteration.

```python
for i in range(5):
    if i == 2:
        continue
    print(i)
```

Output

```
0 1 3 4
```

Use case 🌍

* Skip invalid data
* Ignore unwanted values

---

## `pass`

Does nothing (placeholder).

```python
for i in range(5):
    if i == 3:
        pass
    print(i)
```

Used when **syntax requires a statement but you don't want to write code yet**.

Example 🌍

```python
def future_feature():
    pass
```

---

# 6️⃣ Nested Flow Control

Flow control inside another flow control.

Example:

```python
for i in range(3):
    for j in range(2):
        print(i, j)
```

Output

```
0 0
0 1
1 0
1 1
2 0
2 1
```

### Real-world use 🌍

* Matrix operations
* AI dataset iteration
* Game grids

---

# ⚠️ Common Beginner Mistakes

### ❌ Forgetting colon

Wrong

```python
if x > 10
```

Correct

```python
if x > 10:
```

---

### ❌ Infinite loop

```python
i = 1

while i <= 5:
    print(i)
```

This never stops because `i` is never updated.

Correct

```python
i += 1
```
