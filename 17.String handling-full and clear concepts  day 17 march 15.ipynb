# String handling 

### Why We Need **String Handling** in Python

A **string** is simply **text data** (characters).
Example: `"Hello"`, `"Phanindra"`, `"AI Engineer"`

In real-world programs, **most data is text** (names, emails, passwords, messages, files, etc.).
So we need **string handling** to **create, modify, search, and process text** efficiently.

---

## 1️⃣ To **Store and Process Text Data**

Almost every program deals with text.

**Example:**

```python
name = "Phanindra Reddy"
print("Hello", name)
```

Real-world uses:

* User names
* Addresses
* Messages
* File contents
* Chat applications

---

## 2️⃣ To **Search Information in Text**

We often need to check if some word exists.

**Example**

```python
text = "I am learning Python"

print("Python" in text)
```

Output:

```
True
```

Real-world uses:

* Searching keywords in documents
* Detecting spam messages
* Filtering comments

---

## 3️⃣ To **Modify Text**

Sometimes we need to change text.

**Example**

```python
msg = "hello world"
print(msg.upper())
```

Output

```
HELLO WORLD
```

Real-world uses:

* Capitalizing names
* Formatting messages
* Cleaning user input

---

## 4️⃣ To **Extract Parts of Text**

We may need only a **part of a string**.

**Example**

```python
email = "user@gmail.com"
print(email.split("@")[0])
```

Output

```
user
```

Real-world uses:

* Extract username from email
* Get domain names
* Parse data

---

## 5️⃣ To **Validate User Input**

Programs must check whether input is correct.

**Example**

```python
password = "abc123"

print(password.isalnum())
```

Real-world uses:

* Password validation
* Email validation
* Form validation

---

## 6️⃣ To **Format Output**

Make programs display data nicely.

**Example**

```python
name = "Phanindra"
age = 22

print(f"My name is {name} and I am {age} years old")
```

Output

```
My name is Phanindra and I am 22 years old
```

Most files contain text.

Examples:

### `count()` in Python (String Method)

`count()` is used to **count how many times a substring (character or word) appears in a string**.

---

## Syntax

```python
string.count(substring, start, end)
```

### Parameters

* **substring** → The character or word to count
* **start** *(optional)* → Starting index
* **end** *(optional)* → Ending index

---

# 1️⃣ Basic Example

```python
text = "python programming"

print(text.count("p"))
```

**Output**

```
2
```

Explanation
`p` appears **2 times** in the string.

---

# 2️⃣ Counting a Word

```python
text = "I like python because python is powerful"

print(text.count("python"))
```

**Output**

```
2
```

---

# 3️⃣ Using Start and End Index

```python
text = "banana"

print(text.count("a", 2, 6))
```

**Output**

```
2
```

Explanation
It counts `"a"` from **index 2 to 6**.

---

# 4️⃣ Counting Substring

```python
text = "abababa"

print(text.count("aba"))
```

**Output**

```
2
```

Python counts **non-overlapping occurrences**.

---



# Important Points

✔ Works only on **strings** (also available for lists and tuples).
✔ Returns an **integer**.
✔ **Case-sensitive**.


✅ **Simple Definition**

> `count()` returns the **number of times a character or substring occurs in a string**.

## `index()` in Python (String Method)

`index()` is used to **find the position (index number) of a substring or character in a string**.

It returns the **index of the first occurrence**.

---

## Syntax

```python
string.index(substring, start, end)
```

### Parameters

* **substring** → character or word to search
* **start** *(optional)* → starting position
* **end** *(optional)* → ending position

---

## Example 1: Find a Character

```python
text = "python programming"
print(text.index("p"))
```

**Output**

```
0
```

Explanation:
The first `"p"` is at **index 0**.

---

## Example 2: Find a Word

```python
text = "I like python"
print(text.index("python"))
```

**Output**

```
7
```

Explanation:
`python` starts at **index 7**.

---

## Example 3: Using Start Index

```python
text = "banana"
print(text.index("a", 2))
```

**Output**

```
3
```

Explanation:
Search starts from **index 2**.

---

## Important Behavior

⚠ If the substring **is not found**, `index()` gives an **error**.

Example:

```python
text = "python"
print(text.index("z"))
```

Output

```
ValueError: substring not found
```

---

## Difference Between `index()` and `find()`

| Method    | If substring not found |
| --------- | ---------------------- |
| `index()` | Gives **error**        |
| `find()`  | Returns **-1**         |

Example:

```python
text = "python"

print(text.find("z"))
print(text.index("z"))
```

Output

```
-1
ValueError
```

---

## Real-World Example

Finding position of `"@"` in an email:

```python
email = "user@gmail.com"

print(email.index("@"))
```

Output

```
4
```

This helps extract username or domain.

---

✅ **Simple Definition**

> `index()` returns the **position of the first occurrence of a substring in a string**.

## `split()` in Python

`split()` is a **string method** used to **break a string into a list of words or parts** based on a **separator**.

---

## Syntax

```python
string.split(separator, maxsplit)
```

### Parameters

* **separator** *(optional)* → character used to split the string
* **maxsplit** *(optional)* → maximum number of splits

Default separator = **space**

---

# 1️⃣ Basic Example

```python
text = "python is powerful"
print(text.split())
```

**Output**

```
['python', 'is', 'powerful']
```

Explanation:
The string is split **at spaces**.

---

# 2️⃣ Using a Separator

```python
text = "apple,banana,mango"
print(text.split(","))
```

**Output**

```
['apple', 'banana', 'mango']
```

Explanation:
Split happens at **`,`**.

---

# 3️⃣ Using `maxsplit`

```python
text = "one two three four"
print(text.split(" ", 2))
```

**Output**

```
['one', 'two', 'three four']
```

Explanation:
Only **2 splits** happen.

# 4️⃣ Splitting Characters

```python
text = "a-b-c-d"
print(text.split("-"))
```

**Output**

```
['a', 'b', 'c', 'd']
```

---

# Important Points

✔ Returns a **list**
✔ Default separator is **space**
✔ Used to **extract words or data**

---

# Real-World Example

Extract words from a sentence.

```python
sentence = "I want to become AI engineer"
words = sentence.split()

print(words)
```

Output

```
['I', 'want', 'to', 'become', 'AI', 'engineer']
```

---

# Example: Extract Username from Email

```python
email = "user@gmail.com"
print(email.split("@"))
```

Output

```
['user', 'gmail.com']
```

## A space delimiter is when a space " " is used to separate words or values in a string.

Many delimiters exist in programming, such as:

| Delimiter    | Example                |
| ------------ | ---------------------- |
| Space `" "`  | `"Python AI ML"`       |
| Comma `","`  | `"apple,banana,mango"` |
| Hyphen `"-"` | `"2026-03-15"`         |
| Colon `":"`  | `"12:30:45"`           |


## `join()` in Python

`join()` is a **string method** used to **combine elements of a list, tuple, or iterable into a single string** using a **separator**.

It is the **opposite of `split()`**.

---

## Syntax

```python
separator.join(iterable)
```

* **separator** → character/string used between elements
* **iterable** → list, tuple, or set of strings

---

## Example 1: Join List Elements

```python
words = ["Python", "is", "powerful"]

result = " ".join(words)
print(result)
```

**Output**

```
Python is powerful
```

Explanation:
A **space `" "`** joins the words.

---

## Example 2: Join with `-`

```python
words = ["2026", "03", "15"]

print("-".join(words))
```

**Output**

```
2026-03-15
```

---

## Example 3: Join Characters

```python
letters = ["P", "Y", "T", "H", "O", "N"]

print("".join(letters))
```

**Output**

```
PYTHON
```

Explanation:
Empty string `""` joins characters **without space**.

---

## Example 4: Join Tuple

```python
t = ("apple", "banana", "mango")

print(", ".join(t))
```

**Output**

```
apple, banana, mango
```

---

## Important Rule

⚠ All elements must be **strings**.

Example (Error):

```python
data = ["python", 123]

print(",".join(data))
```

Output:

```
TypeError: sequence item must be str
```

Correct way:

```python
data = ["python", str(123)]

print(",".join(data))
```

---

## Real-World Example

Convert a list of words into a **sentence**.

```python
words = ["I", "want", "to", "become", "an", "AI", "Engineer"]

sentence = " ".join(words)

print(sentence)
```

Output

```
I want to become an AI Engineer
```

---

## `startswith()` in Python

`startswith()` is a **string method** used to check whether a **string begins with a specified substring**.

* Returns **`True`** → if the string starts with the given substring
* Returns **`False`** → otherwise

---

## Syntax

```python
string.startswith(prefix, start, end)
```

### Parameters

* **prefix** → substring to check
* **start** *(optional)* → starting index
* **end** *(optional)* → ending index

---

## Example 1: Basic Example

```python
text = "Python programming"

print(text.startswith("Python"))
```

**Output**

```
True
```

Explanation:
The string **starts with "Python"**.

---

## Example 2

```python
text = "Python programming"

print(text.startswith("program"))
```

**Output**

```
False
```

Explanation:
The string does **not start with "program"**.

---

## Example 3: Using Start Index

```python
text = "Python programming"

print(text.startswith("program", 7))
```

**Output**

```
True
```

Explanation:
Starting from **index 7**, the word `"program"` appears.

---

## Example 4

```python
text = "hello world"

print(text.startswith("hello"))
```

Output

```
True
```

---

## Important Points

✔ Returns **True or False**
✔ Used to check the **beginning of a string**
✔ **Case-sensitive**

Example:

```python
text = "Python"

print(text.startswith("python"))
```

Output

```
False
```

---

## Real-World Example

Check if an email is from **Gmail**:

```python
email = "gmail_user@gmail.com"

if email.startswith("gmail"):
    print("Gmail user")
```

---

## Related Method

| Method         | Purpose                    |
| -------------- | -------------------------- |
| `startswith()` | Checks beginning of string |
| `endswith()`   | Checks end of string       |

Example:

```python
text = "file.txt"

print(text.endswith(".txt"))
```

Output

```
True
```

---

✅ **Simple Definition**

> `startswith()` checks whether a string **begins with a specific substring**.


## `endswith()` in Python

* Returns **`False`** → otherwise

---

## Syntax

```python
string.endswith(suffix, start, end)
```

### Parameters

* **suffix** → substring to check at the end
* **start** *(optional)* → starting index
* **end** *(optional)* → ending index

---

## Example 1: Basic Example

```python
text = "Python programming"

```

**Output**

```
True
```

Explanation:
The string **ends with "programming"**.

---


## `strip()`, `lstrip()`, `rstrip()` in Python

These are **string methods** used to **remove unwanted characters (usually spaces)** from a string.

---

# 1️⃣ `strip()`

`strip()` removes **spaces from both the left and right sides** of a string.

### Syntax

```python
string.strip()
```

### Example

```python
text = "   Python Programming   "
print(text.strip())
```

**Output**

```
Python Programming
```

Explanation
Spaces at the **beginning and end** are removed.

---

# 2️⃣ `lstrip()`

`lstrip()` removes **spaces only from the left side** (beginning) of the string.

### Syntax

```python
string.lstrip()
```

### Example

```python
text = "   Python Programming"
print(text.lstrip())
```

**Output**

```
Python Programming
```

Explanation
Spaces at the **left side** are removed.

---

# 3️⃣ `rstrip()`

`rstrip()` removes **spaces only from the right side** (end) of the string.

### Syntax

```python
string.rstrip()
```

### Example

```python
text = "Python Programming   "
print(text.rstrip())
```

**Output**

```
Python Programming
```

Explanation
Spaces at the **right side** are removed.

---

# Example Showing All Three

```python
text = "   Python   "

print(text.strip())
print(text.lstrip())
print(text.rstrip())
```

Output

```
Python
Python   
   Python
```

---

# Removing Specific Characters

These methods can also remove **specific characters**.

Example:

```python
text = "***Python***"

print(text.strip("*"))
```

Output

```
Python
```

---

# Real-World Example

Cleaning user input:

```python
name = input("Enter your name: ").strip()

print("Hello", name)
```

If user types:

```
   Phanindra   
```

Output:

```

```

---

# Difference Summary

| Method     | Removes             |
| ---------- | ------------------- |
| `strip()`  | Left + Right spaces |
| `lstrip()` | Left spaces only    |
| `rstrip()` | Right spaces only   |

---

✅ **Simple Definition**

* `strip()` → removes spaces from **both sides**
* `lstrip()` → removes spaces from **left side**
* `rstrip()` → removes spaces from **right side**
