# Types of Copy Techniques in Python

=> In Python Programming, we have 2 Types of Copy Techniques. They are:

1. **Shallow Copy**
2. **Deep Copy**

---

## 1. Shallow Copy: Copying the contents of one list object to another list object. 
    - To implement this, we use copy()
    syntax: list_obj2=list_obj1.copy()

---

=> The Properties of Shallow Copy are:

* **a)** The Initial Content of Both the Objects are Same.
* **b)** The Memory Address of Both the Objects are Different.
* **c)** The Modifications are Independent (Whatever the Changes we do on one Object , those changes are not reflecting to another object 
because memory addresses are different ...)




### **2. Deep Copy**

---

**=> The Properties of Deep Copy are**

* **a)** The Initial Content of Both the Objects are Same.
* **b)** The Memory Address of Both the Objects are Same.
* **c)** The Modifications are Dependent.
* (Whatever the Changes we do on one Object, Those Changes are Reflecting to another object bcoz Memory address are Same)



**=> To Implement Deep Copy, we use Assignment Operator ( = ) Only.**

**=> Syntax: Object2 = Object1**


````python
l1=[12,34,True,"India"]
print("Source list is ", l1," and memory address is", id(l1))
l2=l1.copy()
print("Destination list is ", l2," and memory address is", id(l2))
````

````python
l1.append(1000) # changes made in l1 doesn't affect the l2 because memory addresses are different here 
print(l2)
````

````python
l2.append("added")
print(l2)
````

````python
print(l2)
print(l1)
````

````python
l3=[12.45,34.89,True,"Germany"]
print("Source list is ", l3," and memory address is", id(l3))
l4=l3
print("Destination list is ", l4," and memory address is", id(l4))
````

````python
l3.append("added element  ")
print(l3) # changes made in the l3 here 
print(l4) # changes reflected in the l4 also
````

````python
l3.remove(True)
print(l3,id(l3))
print(l4,id(l4))
````

 ## `copy()` in Python

The `copy()` method is mainly used with **lists, dictionaries, and sets** to create a **shallow copy** of the object.

---

# 1️⃣ `copy()` in List

### ✅ Syntax

```python
new_list = old_list.copy()
```

### ✅ Example

```python
a = [10, 20, 30]
b = a.copy()

print(b)          # [10, 20, 30]
print(a is b)     # False (different objects)
```

### 🔎 What happens?

* A **new list object** is created.
* Elements are copied.
* Both lists are **independent objects**.

---

### ⚠️ Important (Shallow Copy Concept)

```python
a = [[1, 2], [3, 4]]
b = a.copy()

b[0][0] = 100

print(a)  # [[100, 2], [3, 4]]
```

❗ Why did `a` change?

Because:

* `copy()` makes a **shallow copy**
* Inner lists are still referencing the same objects

---

# 2️⃣ `copy()` in Dictionary

```python
d1 = {"a": 1, "b": 2}
d2 = d1.copy()

print(d2)
print(d1 is d2)  # False
```

Same rule:

* New dictionary created
* Values are shallow copied

---

# 3️⃣ `copy()` in Set

```python
s1 = {1, 2, 3}
s2 = s1.copy()

print(s2)
```

Creates a new set object.

---

# 🔥 Shallow Copy vs Deep Copy

To create a **deep copy**, we use:

```python
import copy

a = [[1, 2], [3, 4]]
b = copy.deepcopy(a)

b[0][0] = 999

print(a)  # No change
```

### Difference:

| Type         | Copies Outer | Copies Inner Objects |
| ------------ | ------------ | -------------------- |
| `copy()`     | ✅ Yes        | ❌ No                 |
| `deepcopy()` | ✅ Yes        | ✅ Yes                |

---

# 💡 Other Ways to Copy a List

```python
b = a[:]        # slicing
b = list(a)     # constructor
```

---

## `reverse()` in Python (List Method)

### 1️⃣ Definition

`reverse()` is a **list method** used to **reverse the order of elements in a list**.

It **modifies the original list itself**.

So it is called an **in-place operation**. (reversed elements placed in the same object itself).

---

# 2️⃣ Syntax

```python
list.reverse()
```

* No arguments are required.
* It changes the original list.

---

# 3️⃣ Example

```python
a = [10, 20, 30, 40]

a.reverse()

print(a)
```

Output

```
[40, 30, 20, 10]
```

Explanation:

Original list

```
[10, 20, 30, 40]
```

After reverse

```
[40, 30, 20, 10]
```

---

# 4️⃣ Important Concept (Return Value)

`reverse()` **returns None**.

Example:

```python
a = [1,2,3]

b = a.reverse()

print(b)
```

Output

```
None
```

Because the list is already modified.

Correct way:

```python
a.reverse()
print(a)
```

---

# 5️⃣ Reverse vs Slicing

Another way to reverse:

```python
a = [10,20,30]

b = a[::-1]

print(b)
```

Output

```
[30,20,10]
```

Difference:

| Method      | Changes Original List | Creates New List |
| ----------- | --------------------- | ---------------- |
| `reverse()` | ✅ Yes                 | ❌ No             |
| `[::-1]`    | ❌ No                  | ✅ Yes            |

---

# 6️⃣ Internal Idea

If list is

```
[1,2,3,4]
```

`reverse()` swaps elements:

```
1 ↔ 4
2 ↔ 3
```

Final result:

```
[4,3,2,1]
```

Time Complexity

```
O(n)
```

---

# 7️⃣ Practical Use

Used in:

### Data processing

```python
numbers.reverse()
```

### Reversing logs

```python

```

### Algorithms

Many algorithms need **reverse traversal**.

## `sort()` in Python (List Method)

### 1️⃣ Definition

`sort()` is a **list method** used to **arrange elements of a list in a specific order**.

By default, it sorts the list in **ascending order**.

It **modifies the original list** (in-place operation).


| Parameter | Meaning                              |
| --------- | ------------------------------------ |
| `key`     | function used to customize sorting   |
| `reverse` | if `True`, sorts in descending order |

```python
print(a)
```

## `extend()` in Python (List Method)

# 3️⃣ Basic Example

```python
print(a)
```


Explanation
Elements of `b` are **added individually** to `a`.

---

```python
print(a)
```

```

```



---

# 6️⃣ Difference Between `append()` and `extend()`

### append()

```python
a = [1,2,3]
a.append([4,5])

print(a)
```

Output

```
[1,2,3,[4,5]]
```

The **entire list becomes one element**.

---

### extend()

```python

```

Output


Elements are added **individually**.

---

### Comparison Table

| Method     | Adds                 | Result          |
| ---------- | -------------------- | --------------- |

# write a program to find out the maximum number of occurences of the element


````python
l1=[23,45,54,4,45,32,45,76,89,45,90,-54,32,45] # 45 occured more number of times
max(l1,key=l1.count)
````


## 1️⃣ What this code does

It finds the **element that appears the most number of times in the list**.

So it returns the **most frequent element**.

---

# 2️⃣ Step-by-step Understanding

### List

```
[23,45,54,4,45,32,45,76,89,45,90,-54,32,45]
```

Occurrences:

| Element | Count |
| ------- | ----- |
| 23      | 1     |
| 45      | 5     |
| 54      | 1     |
| 4       | 1     |
| 32      | 2     |
| 76      | 1     |
| 89      | 1     |
| 90      | 1     |
| -54     | 1     |

So the **maximum occurrence = 45 (5 times)**.

---

# 3️⃣ How `max()` Works Here

Syntax used:

```
max(iterable, key=function)
```

Here

```
iterable → l1
key → l1.count
```

So Python checks:

```
l1.count(23) = 1
l1.count(45) = 5
l1.count(54) = 1
l1.count(4)  = 1
l1.count(32) = 2
...
```

Python selects the element with **highest count**.

Result:

```
45
```

---

# 4️⃣ Internal Logic

Python is doing something like this internally:

```
max_value = None
max_count = 0

for x in l1:
    c = l1.count(x)
    if c > max_count:
        max_count = c
        max_value = x
```

Final answer

```
45
```

---

# 5️⃣ Output

```
45
```

---

# 6️⃣ Important Concept

This method works but is **not efficient for very large lists**.

Time Complexity:

```
O(n²)
```

Because `count()` runs for every element.

---

# 7️⃣ Better Way (Professional Python)

Using dictionary:

```python
from collections import Counter

l1=[23,45,54,4,45,32,45,76,89,45,90,-54,32,45]

print(Counter(l1).most_common(1))
```

Output

```
[(45,5)]
```

---

✅ **Key idea**

```
max(l1, key=l1.count)
```

means

```
find element whose count is maximum
```


## **Inner OR Nested List**

============================================================

**=> The Process of Defining One List inside of another list is called Inner OR Nested List.**

**=> Syntax:** `listobj = [Val1, Val2,....[val11, val12,....Val1n], [Val21, Val22,....Val2n],......Val-n]`

* Here **Val1, Val2.....Val-n** are called **Outer List Elements**.
* Here **val11, val12...val1n** are called **One inner Elements**.
* Here **val21, val22...val2n** are called **another inner Elements**.

**=> On Inner List, we can perform Both Indexing and Slicing Operations.**
**=> On Inner List, we can apply all types of Pre-defined functions of list.**
