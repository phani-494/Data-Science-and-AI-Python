
# 🧠 1. What is `filter()` (Real Meaning)

`filter()` is used to **select elements** from a collection based on a condition.

👉 It **keeps only those elements** for which the condition is `True`.

---

## 🔹 Syntax

```python
filter(function, iterable)
```

* `function` → decides True/False
* `iterable` → list, tuple, etc.
* Returns → **filter object** (so we convert to list)

---

# ⚡ 2. Basic Example

```python
nums = [1, 2, 3, 4, 5, 6]

even = list(filter(lambda x: x % 2 == 0, nums))
print(even)
```

👉 Output:

```
[2, 4, 6]
```

---

# 🔥 3. Real-World Thinking

👉 Think of `filter()` like a **sieve**:

* Input → all data
* Output → only valid data

# `filter()` in Python

## What is it?
`filter()` is a built-in function that **filters items from a list** (or any iterable) based on a condition.

---

## Syntax
```python
filter(function, iterable)
```
- **function** → returns `True` or `False` for each item
- **iterable** → list, tuple, etc. to filter from

> Only items where the function returns `True` are kept.

---

## Simple Example

```python
numbers = [1, 2, 3, 4, 5, 6]

evens = list(filter(lambda x: x % 2 == 0, numbers))

print(evens)  # Output: [2, 4, 6]
```

---

## Using a Regular Function

```python
def is_positive(n):
    return n > 0

numbers = [-3, -1, 0, 2, 5]
result = list(filter(is_positive, numbers))

print(result)  # Output: [2, 5]
```

---

## Common Examples

### 1. Filter even numbers
```python
nums = [1, 2, 3, 4, 5, 6]
evens = list(filter(lambda x: x % 2 == 0, nums))
# Output: [2, 4, 6]
```

### 2. Filter words by length
```python
words = ["hi", "hello", "hey", "howdy"]
long_words = list(filter(lambda w: len(w) > 3, words))
# Output: ['hello', 'howdy']
```

### 3. Filter non-empty strings
```python
names = ["Alice", "", "Bob", "", "Charlie"]
valid = list(filter(None, names))  # None removes falsy values
# Output: ['Alice', 'Bob', 'Charlie']
```

### 4. Filter students who passed
```python
students = [("Alice", 85), ("Bob", 40), ("Charlie", 72), ("Dave", 35)]
passed = list(filter(lambda s: s[1] >= 50, students))
# Output: [('Alice', 85), ('Charlie', 72)]
```

---

## `filter()` vs List Comprehension

```python
nums = [1, 2, 3, 4, 5, 6]

# Using filter
evens = list(filter(lambda x: x % 2 == 0, nums))

# Using list comprehension (more Pythonic)
evens = [x for x in nums if x % 2 == 0]

# Both give: [2, 4, 6]
```

---

## Key Points

| Point | Detail |
|-------|--------|
| Returns | A `filter` object (convert with `list()`) |
| Keeps | Only items where function returns `True` |
| `None` as function | Removes all falsy values (0, "", None, False) |
| Alternative | List comprehension is often preferred |

---

## Quick Summary

> `filter()` = **go through each item → keep it only if condition is True**

### ** map()**
============================================================

* `map()` is used for obtaining a new iterable object from an existing iterable object by applying old iterable elements to the function.
* In other words, `map()` is used for obtaining a new list of elements from an existing list of elements by applying old list elements to the function.

**Syntax:**
`varname = map(FunctionName, Iterable_object)`

* Here `varname` is an object of type `<class, 'map'>` and we can convert it into any iterable object by using type casting functions.
* `FunctionName` represents either normal functions or anonymous functions.
* `Iterable_object` represents Sequence, List, Set, and Dict types.
* The execution process of `map()` is that "`map()` sends every element of an iterable object to the specified function, processes it, and returns the modified value (result), and a new list of elements will be obtained." This process will be continued until all elements of the `Iterable_object` are completed.

# `map()` in Python

## What is it?
`map()` is a built-in function that **applies a function to every item** in an iterable and returns the transformed result.

---

## Syntax
```python
map(function, iterable)
```
- **function** → operation to apply on each item
- **iterable** → list, tuple, etc. to work on

> Every item gets transformed, nothing is removed.

---

## Simple Example

```python
numbers = [1, 2, 3, 4, 5]

squares = list(map(lambda x: x**2, numbers))

print(squares)  # Output: [1, 4, 9, 16, 25]
```

---

## Using a Regular Function

```python
def double(n):
    return n * 2

numbers = [1, 2, 3, 4]
result = list(map(double, numbers))

print(result)  # Output: [2, 4, 6, 8]
```

---

## Common Examples

### 1. Square all numbers
```python
nums = [1, 2, 3, 4, 5]
squares = list(map(lambda x: x**2, nums))
# Output: [1, 4, 9, 16, 25]
```

### 2. Convert strings to uppercase
```python

# Output: ['ALICE', 'BOB', 'CHARLIE']
```

### 3. Convert strings to integers

```python
# Output: [1, 2, 3, 4]
```

### 4. Add tax to prices
```python

with_tax = list(map(lambda p: p * 1.18, prices))
# Output: [118.0, 236.0, 354.0]
```

---

```python

```

---

## `map()` vs List Comprehension

```python
nums = [1, 2, 3, 4, 5]

# Using map
squares = list(map(lambda x: x**2, nums))

# Using list comprehension (more Pythonic)

# Both give: [1, 4, 9, 16, 25]
```

---

## `map()` vs `filter()`

| Feature | `map()` | `filter()` |
|--------|---------|-----------|
| Purpose | **Transform** every item | **Select** certain items |
| Output size | Same as input | Same or smaller |
| Function returns | Transformed value | `True` or `False` |
| Example | Square all numbers | Keep only even numbers |

---

## Combining `map()` and `filter()`

```python
numbers = [1, 2, 3, 4, 5, 6]

# Filter evens, then square them
result = list(map(lambda x: x**2, filter(lambda x: x % 2 == 0, numbers)))

print(result)  # Output: [4, 16, 36]
```

---

## Key Points

| Point | Detail |
|-------|--------|
| Returns | A `map` object (convert with `list()`) |
| Transforms | Every single item in the iterable |
| Multiple iterables | Supported — stops at shortest one |
| Alternative | List comprehension is often preferred |

---

## Quick Summary

> `map()` = **go through each item → transform it → return new collection**
