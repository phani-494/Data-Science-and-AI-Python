## Write a program to print the factorial of a number without using the modules 

```python
a =int(input("Enter the Number"))
fac=1
i=1
while(i<=a):
    fac=fac*i
    i=i+1
print(fac)
    
```

```python
n = int(input("Enter a number: "))
fact = 1

while n > 0:
    fact *= n
    n -= 1

print("Factorial =", fact)
```

## Garbage Collector in Python

The garbage collector (GC) in Python is an automatic memory management system that frees memory occupied by objects that are no longer in use.

You don’t need to manually delete memory like in C or C++.

####  Why Garbage Collection Is Needed

When objects are created, they occupy memory.
If unused objects are not removed:

❌ Memory leaks occur

❌ Program slows down

## Python’s GC automatically:

Detects unused objects

Reclaims memory
