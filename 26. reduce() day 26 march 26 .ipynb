## 3. reduce()
==================================================

* **`reduce()`** is used for obtaining a single element / result from a given iterable object by applying to a function.
* **Syntax:-**
    `varname = reduce(function-name, iterable-object)`
* Here **varname** is an object of **int, float, bool, complex, str** only.
* The **`reduce()`** belongs to a pre-defined module called **"functools"**.

--------------------------------------------------

### Internal Flow of reduce()
--------------------------------------------------

* **Step-1:-** Initially, `reduce()` selects **First Two** values of Iterable object and places them in **First var** and **Second var**.
* **Step-2:-** The **function-name** (lambda or normal function) utilizes the values of **First var** and **Second var** and applies to the specified logic and obtains the result.
* **Step-3:-** `reduce()` places the result of function-name in **First variable** and `reduce()` selects the succeeding element of Iterable object and places in **second variable**.
* **Step-4:-** Repeat **Step-2** and **Step-3** until all elements are completed in the Iterable object and returns the result of **First Variable**.

--------------------------------------------------

## 🔥 What `reduce()` actually does

`reduce()` takes:

1. A function (here: `lambda x, y: x + y`)
2. An iterable (your list)

👉 It **reduces the entire list to a single value** by applying the function repeatedly.

---

```

Step 3:
x = 6, y = 4 → 10

Step 4:
x = 10, y = 5 → 15

Step 5:
x = 15, y = 6 → 21

Step 6:
x = 21, y = 9 → 30
```

👉 Final result = `30`


* `iterable` → list, tuple, etc.
* `initial_value` (optional)

# 🧠 Internal Working (Very Important)

`reduce()` internally behaves like this:

```python
def my_reduce(function, iterable):
    result = iterable[0]

    for element in iterable[1:]:
        result = function(result, element)

    return result
```

## `reduce()` in Python

`reduce()` applies a function **cumulatively** to items in a sequence, reducing it to a single value.

**Import it first** — it lives in `functools`:

```python
from functools import reduce
```

---

### How it works

```
[a, b, c, d]  →  f(f(f(a, b), c), d)  →  single value
```

Each step takes the **accumulated result** and the **next element**, feeding the output forward.

---

### Syntax

```python
reduce(function, iterable, initializer=None)
```

| Parameter | Description |
|---|---|
| `function` | Takes 2 args: accumulator + current item |
| `iterable` | The sequence to reduce |
| `initializer` | Optional starting value |

---

### Examples

**Sum of a list:**
```python
reduce(lambda acc, x: acc + x, [1, 2, 3, 4, 5])
# 1→3→6→10→15  ✓  returns 15
```

**Find the maximum:**
```python
reduce(lambda acc, x: acc if acc > x else x, [3, 1, 4, 1, 5, 9])
# returns 9
```

**Flatten a nested list:**
```python
reduce(lambda acc, x: acc + x, [[1, 2], [3, 4], [5, 6]])
# returns [1, 2, 3, 4, 5, 6]
```

**Using an initializer:**
```python
reduce(lambda acc, x: acc + x, [1, 2, 3], 100)
# starts at 100 → 101 → 103 → 106  ✓  returns 106
```

---

### When to use (and when not to)

✅ **Good for:** custom aggregations, chaining operations, functional-style pipelines

❌ **Prefer built-ins when possible** — they're faster and more readable:

```python
# Instead of reduce for common tasks:
sum([1, 2, 3, 4])          # not reduce
max([3, 1, 4, 1, 5])       # not reduce
"".join(["a", "b", "c"])   # not reduce
```

> **Rule of thumb:** if Python has a built-in for it, use that. Reach for `reduce` when your aggregation logic is custom.
