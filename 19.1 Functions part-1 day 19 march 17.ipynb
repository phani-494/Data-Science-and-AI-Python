# 🆚 Structured vs Unstructured Programming

## 🔹 1️⃣ Structured Programming

**Definition:**
A programming approach where code is written in a **clear, logical structure** using:

* sequence
* selection (if-else)
* iteration (loops)

👉 It avoids jumping randomly in code.

### ✅ Features

* Easy to read and understand
* Uses functions, loops, conditions
* Modular (divided into small parts)
* Easy debugging and maintenance

### ✅ Example (Python)

```python
for i in range(1,6):
    if i % 2 == 0:
        print(i)
```

👉 Flow is **clear and step-by-step**

---

## 🔹 2️⃣ Unstructured Programming

**Definition:**
A programming style where the flow of control is **irregular**, often using jumps like `goto`.

👉 Code execution is **not in a proper order**.

### ❌ Features

* Hard to read
* Difficult to debug
* No proper structure
* Uses jumps (`goto` statements)

### ❌ Example (Conceptual)

```c
start:
printf("Hello");
goto start;
```

👉 Flow jumps back and forth → confusing

---

# 🔥 Key Differences

| Feature            | Structured Programming | Unstructured Programming  |
| ------------------ | ---------------------- | ------------------------- |
| Code flow          | Sequential & logical   | Random jumps              |
| Readability        | Easy                   | Difficult                 |
| Debugging          | Easy                   | Hard                      |
| Reusability        | High                   | Low                       |
| Control statements | if, for, while         | goto                      |
| Example languages  | Python, C, Java        | Old assembly-style coding |



### **Syntax for Defining a Function in Python**

```python
def functionname(list of formal Params if any): <--- Function Heading
    """doc string """
    statement-1
    statement-2                                <--- Function Body
    -----------
    statement-n
```

> **Note:** Function def = **Function Heading** + **Function Body**

---

### **Explanation:**

1.  **`def`** is a keyword used for defining **Programmer-defined Functions**.
2.  **`functionname`** represents a valid variable name and is treated as the function name. Every function name is an object of type `<class, 'function'>`.
3.  **`list of formal params`** represents a list of variable names used in the function heading. They are used for **storing or holding the input(s)** coming from function call(s).
4.  **`"""doc string """`** represents the documentation string. It is used for giving or writing the **description about the functionality** of the function.
5.  **`statement-1, statement-2...statement-n`** is an indentation block of statements. This block processes the input or logic for problem-solving; it is known as **Business Logic** or **Func Proc Logic**.
6.  In the **Function Body**, we use some special variables called **Local Variables**, whose purpose is to store temporary results.
7.  **The values of Formal Params and Local Variables can be accessed only inside corresponding Function Definition but not possible to access in other part of the program and in Other Function Definitions (Scope of Formal params and Local Var)**

### **Quick Technical Insight**

* **PVM (Python Virtual Machine):** As your note mentions, the PVM is the "engine" that runs your code. It loads the function into memory when it sees the `def` keyword, but it stays "idle" until it encounters a function call.
* **The Three Phases:** This is often referred to as the **IPO model** (Input-Process-Output).


**Example matching your phases:**
```python
# 1. Defining the Function (Heading & Body)
def calculate_square(number): # INPUT (Formal Parameter)
    result = number * number  # PROCESS (Business Logic)
    return result             # OUTPUT (Return Statement)

# 2. Function Call (The only way to make PVM execute the logic)
print(calculate_square(5)) 
```

# define a function to add 2 numbers using functions 

### Approach-1 

````python
def add_op(a,b): # a,b are formal parameters  # function definition
    """ This function performs addition of 2 numbers"""
    c=a+b # c is local variable 
    return a+b # returns the local variable logic expression or results to the function call
res=add_op(int(input("Enter a:")),int(input("Enter b:"))) # function call [must be defined after the function definition]
print(res)
# function  execution starts from the line 5 here and once it encounters function call then it goes to function definition stored in memory
````

📌 Final Flow Summary

👉 Order of execution:

Function is defined (stored, not executed)

Function call is reached

Input is taken

Values passed to function

Function runs

Result returned



# 🔥 Step-by-Step Explanation

```python

```

* `c` is a **local variable**
* It exists **only inside the function**
* It cannot be accessed outside

👉 Important:
You calculated `c`, but you are **not using it in return**

```python

```

* This sends the result back to the caller
