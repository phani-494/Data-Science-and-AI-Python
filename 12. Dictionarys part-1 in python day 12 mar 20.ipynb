# Dictionaries 

## Dict Category Data Type

* **The data Type in Dict Category** is 'dict'.
* **dict** is one of the pre-defined class.
* The purpose of dict data type is that "To Store the Data in the form of (Key, Value)".
* In (Key, value), the Values of Key Represents unique values and Values of Value Represents May or May not be Unique.
* To store (Key, value) in dict, we use curly braces and (Key, value) separated by comma.
* **Syntax:** `varname={Key1:Val1, Key2:Val2,........,Key-n:Val-n}`
* here Key1, Key2...Key-n Represents of Values of Key
* Val1, Val2,....Val-n Represents of Values of Value


* An object of dict maintains **Insertion Order**.
* An object of dict **never Contains Indices**. So that we can't perform Both Index and Slicing Operations.
* An object of dict belongs to **MUTABLE**. In detail, the values of Key belongs to **IMMUTABLE** and Values of Value belongs to **MUTABLE**.
* In Python Programming, we can create Two Types of Dict Objects.
* They are:
a) Empty Dict

b) Non-Empty Dict

---



- It contains a key and related value for that key.
- examples are 1.  Account_no(Key) : Account_balance (value)
- Contact_Name (key): Phone_no (value)
- halltkt_no(key): rank(value)
- employee_id(key): emp_name(value)

## 📘 Dictionaries in Python

A **dictionary** in Python is a **data structure that stores data in `key : value` pairs**.

It is used when you want to **map one piece of data to another**.

Example:

* Student → Marks
* Product → Price
* Country → Capital

So instead of storing values by **index (like lists)**, dictionaries store values by **keys**.

---

# 1️⃣ Basic Syntax

```python
d = {key1:value1, key2:value2, key3:value3}
```

Example:

```python
student = {
    "name": "Phanindra",
    "age": 22,
    "course": "AI"
}

print(student)
```

Output

```
{'name': 'Phanindra', 'age': 22, 'course': 'AI'}
```

---

# 2️⃣ Important Properties of Dictionaries

| Property              | Explanation                |
| --------------------- | -------------------------- |
| **Key-Value Pair**    | Data stored as `key:value` |
| **Mutable**           | You can change values      |
| **Keys are unique**   | Duplicate keys not allowed |
| **Values can repeat** | Same values allowed        |
| **Unordered**         | No index positions         |
| **Dynamic size**      | Can grow or shrink         |

---

## clear()
- syntax: dictobj.clear()
- used to clear all key -value pairs in the dictionary and gives the empty dictionary 
- if we apply on the empty dictionary, then it gives result as None

````python
print(d2)
d2.clear()
print(d2)
````

````python
# if we apply on the empty dict, then it gives None
print({}.clear())
print(dict().clear())
````

## pop()
- used to remove the key-value pair  using the key.(from the non empty dictobj)
- we should pass the key here else it gives keyerror here
- if we try to pop from the empty dictionary then we get the key error here.

````python
d1={123:"Indra",124:"Mahi",125:"Raja",126:"Pratap"}
print(d1)
print(d1.pop(123))
print(d1.pop(124))
print(d1)
print(d1.pop(500)) # gives error here as 500 key is not present
````

````python
{}.pop(90) # on empty dict, we cannot pop as no elements are present in it 
````

## popitem()
- used to remove the last key-value pair from the non empty dictobj. 
- No need to pass any key here 
- if we try to popitem from the empty dictionary then we get the key error here.

````python
d2={123:"Indra",124:"Mahi",125:"Raja",126:"Pratap"}
print(d2)
print(d2.popitem()) # last pair deleted 
print(d2.popitem())
print(d2)
print(d2.popitem())
````

````python
{}.popitem() # gives key error as dict is empty .
````

## copy()
- this function is used to copy the contents of one object to another dict_obj (shallow copy)


````python
d1={123:"Indra",124:"Mahi",125:"Raja",126:"Pratap"}
print(d1,id(d1))
d2=d1.copy() # shallow copy 
print(d2,id(d2))
````

````python
# make some changes to both dict here and check their address 
d1[12]="Mahesh"
d2["Name"]="Jaski"
print(d1, id(d1))
print(d2, id(d2))
````

````python
d1={123:"Indra",124:"Mahi",125:"Raja",126:"Pratap"}
d2=d1 # deep copy . so changes done in both dict and same memory address here 
d1[12]="Mahesh singh "
d2["Name"]="Jaski reddy "
print(d1, id(d1))
print(d2, id(d2))
````

## 📘 Predefined (Built-in) Functions for **Dictionaries in Python**

Dictionaries have **built-in methods (predefined functions)** that help us **access, modify, and manage key-value pairs** easily.

Syntax of dictionary:

```python

```

Example:


# 1️⃣ `keys()`

Returns **all keys** of the dictionary.

```python
d = {"name":"Phanindra","age":22,"course":"AI"}

print(d.keys())
```

Output

```
dict_keys(['name', 'age', 'course'])
```

✔ Used when you only need **keys**.

---

# 2️⃣ `values()`

Returns **all values**.

```python
print(d.values())
```

Output

```
dict_values(['Phanindra', 22, 'AI'])
```

✔ Useful in **data analysis**.

---

# 3️⃣ `items()`

Returns **key-value pairs** as tuples.

```python
print(d.items())
```

Output

```
dict_items([('name','Phanindra'), ('age',22), ('course','AI')])
```

Very useful for **looping**.

Example:

```python
for k,v in d.items():
    print(k,v)
```

Output

```
name Phanindra
age 22
course AI
```

---

# 4️⃣ `get()`

Safely gets a value using key.

```python

```

Output

```

```

Difference:

```
d["salary"]  ❌ error
```

# 🔟 `setdefault()`

Returns value of key.

If key doesn't exist → creates key.

```python

```


# 1️⃣1️⃣ `fromkeys()`
